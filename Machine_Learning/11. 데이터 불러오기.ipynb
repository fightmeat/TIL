# 파이썬을 통한 XML 파일 다루는 작업 # 동적인 html형태
- eXtensible Markup Language : XML
- HyperText Markup Language : HTML

```python
x_str = """
<book>
    <name> 코딩도장 </name>
    <author> 남재윤 </author>
    <year> 2022 </year>
</book>
"""
```

```python
import xml.etree.ElementTree as et
# 문자열을 XML로 변환하는 함수 fromstring()
book = et.fromstring(x_str)
print(type(book))
```

```python
# root tag = 부모태그
book.tag
```

```python
book_childs = list(book)
print(book_childs)
```

```python
# unpacking : 하나의 변수로 묶여있는 값을 각각의 변수로 변환하는 작업
name, author, year = book_childs
# text 속성이 각 태그에 해당하는 문자를 치환한다.
print(name.text)
print(author.text)
print(year.text)
```

```python
# findtext('tag이름')
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')
print(name,author,year)
```

```python
x2_str = """
<books>
    <book>
        <name> 코딩도장 </name>
        <author> 남재윤 </author>
        <year> 2022 </year>
    </book>
    <book>
        <name> 데이터 시각화 </name>
        <author> 설지윤 </author>
        <year> 2023 </year>
    </book>
</books>
"""
```

```python
books = et.fromstring(x2_str)
print(books.tag)
```

```python
# 여러 개의 자식태그를 확인하여 읽어오는 작업
# 모든 자식 객체를 반환하는 메서드 findall()
for book in books.findall('book'):
    name = book.findtext('name')
    author = book.findtext('author')
    year = book.findtext('year')
    print(name,author,year, '\n')
```

```python
# requests 패키지는 HTML을 가져온다.
import requests
# 무라카미 하루키 '도시와 그 불확실한 벽'의 ISBN
isbn = 9788954699075
url = 'https://www.yes24.com/product/search?query={}'
# url = 'https://www.yes24.com/Product/Goods/122090075'
r = requests.get(url.format(isbn))
# r = requests.get(url)
```

```python
print(r.text)
```

```python
# HTML에서 데이터 추출 : Beautiful Soup(뷰티풀수프)
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
# 태그 위치 찾기 : find('찾을 태그명', attrs=['id':'search']), attributes
pre_link = soup.find('a',attrs={'class':'gd_name'})
```

```python
print(pre_link)
```

```python
print(pre_link['href'])
```

```python
url = 'https://www.yes24.com' + pre_link['href']
r = requests.get(url)
```

```python
print(r.text)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
detail = soup.find('div', attrs={'id':'infoset_specific'})
print(detail)
```

```python
# 테이블 태그를 리스트로 가져오는 작업 : find_all()
tr_list = detail.find_all('tr')
print(tr_list)
```

```python
# 태그안에 텍스트를 가져오는 작업 : get_text()
for tr in tr_list:
    if tr.find('th').get_text() == '쪽수, 무게, 크기':
        data = tr.find('td').get_text()
        break
```
