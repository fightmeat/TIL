* **kind="kde"**

  kind ÌååÎùºÎØ∏ÌÑ∞Ïóê "kde"Î•º ÎåÄÏûÖÌïòÎ©¥ Ïª§ÎÑê Î∞ÄÎèÑ Ï∂îÏ†ïÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏùºÎ≥ÄÎüâ ÎòêÎäî Ïù¥Î≥ÄÎüâ Î∂ÑÌè¨Î•º ÌîåÎ°ØÌåÖÌï©ÎãàÎã§.

* **kind="ecdf"**

  kind ÌååÎùºÎØ∏ÌÑ∞Ïóê "ecdf"Î•º ÎåÄÏûÖÌïòÎ©¥ Í≤ΩÌóòÏ†Å ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®ÏàòÎ•º ÌîåÎ°ØÌåÖÌï©ÎãàÎã§.

* **kde=True**

  kde ÌååÎùºÎØ∏ÌÑ∞Ïóê TrueÎ•º ÎåÄÏûÖÌïòÎ©¥ kde Í≥°ÏÑ†Ïù¥ Ï∂îÍ∞ÄÎêú ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Í∑∏Î¶¥ Ïàò ÏûàÏäµÎãàÎã§.

  hue ÌååÎùºÎØ∏ÌÑ∞Ïóê "species", kind ÌååÎùºÎØ∏ÌÑ∞Ïóê "kde"Î•º ÎåÄÏûÖÌïòÎ©¥, Í∞Å speciesÏùò kde ÌîåÎ°ØÏùÑ Îã§Î•∏ ÏÉâÏÉÅÏúºÎ°ú ÌïòÏó¨ Î≥ÑÎèÑÎ°ú Í∑∏Î¶¥ Ïàò ÏûàÏäµÎãàÎã§.

  hue ÌååÎùºÎØ∏ÌÑ∞Ïóê "species", multiple ÌååÎùºÎØ∏ÌÑ∞Ïóê "stack"ÏùÑ ÎåÄÏûÖÌïòÎ©¥, Í∞Å speciesÏùò ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Îã§Î•∏ ÏÉâÏÉÅÏúºÎ°ú ÌïòÏó¨ ÌïòÎÇòÏùò ÎßâÎåÄÏóê Ïó¨Îü¨ Í∞úÏùò ÎßâÎåÄÎ•º ÏåìÎäî ÌòïÏãùÏúºÎ°ú Í∑∏Î¶¥ Ïàò ÏûàÏäµÎãàÎã§.
