# **반복문(Loop)**





#    **for loop**

반복의 범위가 고정된 반복문






*   for x in 리스트 형태
*   리스트 안의 원소를 x에 모두 할당

```python
# for loop: 반복의 범위 fixed!
for x in [1, 2, 3, 4, 5]:
    print(x)

for st in ['abc', [1, 2], 3.14]:   #다양한 자료형 출력 가능
    print(st)
```

*   **합 구하기**

    for문을 사용하여 합을 구하는 방법에 대해 알아보자


```python
#합 구하기

##숫자로 범위
value_sum = 0
for x in [1, 2, 3, 4, 5]:
    value_sum = value_sum + x
print(value_sum)


##리스트 안에 실수값을 넣어 변수로 범위
data = [1, 1, 2, 3, 4, 5, 8]


value_sum = 0  #초기값
for value in data:      
    value_sum = value_sum + value
print(value_sum)


value_sum = 0  #초기값
for value in data:
    value_sum += value   ##증가 연산자: +=, *=, /=
print(value_sum)
```

#    **range**

원하는 숫자를 만들 수 있는 함수







    




    
*   range(시작, 끝, 증가/감소)으로 작성, 끝값 전까지 영향O (끝값은 해당X)

*   range 안에 인수 1개 : 끝 정해짐, 시작은 0부터

*   range 안에 인수 2개 : 시작과 끝 정해짐

```python
#range(시작, 끝, 증/감), 끝값 전까지 출력(끝값은 해당X)
for i in range(0, 5, 1):
    print(i)

##인수 1개 : 끝 정해짐, 시작은 0부터
##인수 2개 : 시작과 끝 정해짐
for i in range(5):      
    print(i)
    
for i in range(5, 0, -1): 
    print(i)

for i in range(4, -1, -1):
    print(i)

value_sum = 0
for i in range(len(data)):   #리스트의 길이를 측정해 범위 값으로
    value_sum += data[i]     #i번째 리스트의 원소를 더해줌
print(value_sum)
```

# **enumerate**

리스트 안에 원소와 그 값의 순서(인덱스)를 같이 리턴할 수 있는 함수

```python
#enumerate
for i, value in enumerate(data):
    print(i, value)


name = 'dacon'

for i, ch in enumerate(name):
    print(i, ch)
```

# **다중(중첩) for loop**

for문 안에 for문이 중첩적으로 들어있는 for문

```python
#중첩 for문 (nested loop)
for i in [0, 1, 2]:
    for j in ['a', 'b', 'c']:
        print(i, j)

##for문 안에 if문도 가능
value_sum = 0
for value in data:
    if value <= 5:
        value_sum += value
print(value_sum)
```

# **반복 제어**

반복문 실행 중에 제어할 수 있는 키워드



*  **break**

   반복문 **실행을 중단**시키고 싶을 때 사용하는 키워드




```python
#반복제어 : break, continue

#break
value_sum = 0

for value in data:
    value_sum += value
    if value_sum >= 10:  #10보다 커지면 반복문 중단
        break
print(value_sum)
```

ㅤ



*   **continue**

   **실행하지 않고 다음으로 넘어가**고 싶을 때 사용하는 키워드 


```python
#continue
value_sum = 0
for value in data:
    if value < 2:       #2보다 작을 때는 더하지 않고 넘어감
        continue
    value_sum += value
print(value_sum)
```

# **while loop**

**비교연산이 참인 동안** 계속 실행하는 반복문

```python
""" while loop : 비교연산이 참인 동안 """

value_sum = 0
i = 0

while value_sum <= 10:
    value_sum += data[i]
    i += 1 # index
   
print(value_sum)


# 10 ==> 1010 : //(몫), %(나머지), 10진수를 2진수로 바꾸기
decimal = 10
result = str()    #숫자형 => 문자열로 바꾸기

while decimal > 0:
    remainder = decimal % 2
    result = str(remainder) + result    
    decimal = decimal // 2
print(result)   
```
