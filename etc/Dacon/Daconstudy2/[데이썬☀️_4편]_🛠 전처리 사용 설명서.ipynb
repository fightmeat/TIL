### **1) 데이터 전처리란?**

데이터가 가지고 있는 **오류 사항을 점검**하고 **데이터의 구조 및 특성**을 우리의 쓰임에 맞춰 용이하게 **수정하는 과정**입니다.
<br><br>

### **2) 전처리의 필요성**



**[이상값, 결측값의 영향]**
  
*   이상값이 존재한다면 **데이터가 불완전**하기 때문에 오차 분산을 증가시키고 추정치가 편향되는 등 데이터 분석의 **결과 자체에 영향**을 주게 됩니다.

*   유의미한 결과를 얻어내기 위해서는 이상값과 결측값을 **제거 및 대체하는 과정**이 필요합니다.

**[데이터 파악 및 가공]**

*   본격적으로 분석에 들어가기 앞서 먼저 결측치나 이상값이 존재하는지, 데이터의 단위가 잘못되어 있는지 등 **데이터의 상태와 가치에 대해 파악**하는 과정이 필요합니다. 

*   우리가 해당 **데이터를 사용하려는 목적**에 맞춰 '범주형 데이터를 숫자로 변경'하거나, '수치 데이터의 분포를 정규화'하는 등의 과정이 필요합니다.

## **3. 데이터 삭제 : Drop**

DataFrame에서 데이터를 삭제할 때 drop() 메소드를 사용합니다.

drop()의 원형은 다음과 같습니다.

DataFrame(변수명).drop(**labels=None**, **axis=0**, index=None, columns=None, level-=None, **inplace=False**, errors='raise')
<br><br>

가장 중요한 파라미터는 **labels, axis, inplace**입니다.

*  axis = 0 : 인덱스 방향 축, axis = 1 : 칼럼 방향 축을 지정합니다.

*  만약 axis = 1로 칼럼을 지정하게 된다면 이때 labels에 원하는 칼럼 명을 입력하여 그에 해당하는 칼럼이 드롭하게 됩니다. 반대로 axis = 0으로 인덱스를 지정하고 labels에 원하는 인덱스 명을 입력하면 그에 해당하는 인덱스가 드롭하게 됩니다.

*  일반적으로 기존 칼럼 값을 가공해 새로운 칼럼을 생성한 후 기존 칼럼 값을 삭제하는 경우가 많아 axis = 1 설정하는 경우가 많습니다.




ㅤ

앞서 사용한 penguin 데이터를 사용하여 실습을 진행하겠습니다.ㅤ

```python
# axis = 1
penguin_df_drop = penguin_df.drop('Island', axis = 1)
penguin_df_drop.head(3)
```

axis = 1을 입력하고 labels에 칼럼 명인 Island를 입력하여 penguin_df_drop 변수에 드랍된 결과값을 할당합니다.

head() 메소드를 이용하여 Island 칼럼 상위 3개 데이터를 확인하면 Island 칼럼이 없어진 것을 확인할 수 있습니다.

ㅤ

```python
# axis = 0
penguin_df_drop = penguin_df.drop([0, 1, 2], axis = 0)
penguin_df_drop.head(3)
```

axis = 0를 입력하고 labels에 인덱스 명인 [0, 1, 2]를 입력하여 penguin_df_drop 변수에 드랍된 결과값을 할당합니다.

head() 메소드로 상위 3개 데이터를 확인하면 인덱스가 0, 1, 2는 드랍되고 3부터 있는 것을 확인할 수 있습니다. 

ㅤ

*  inplace=False(디폴트값)이면 원본 DataFrame은 유지하고 드롭된 DataFrame을 새롭게 객체 변수로 받을 수 있습니다.

```python
penguin_df.head(3)
```

## **4. 데이터 병합 : Merge**

https://mizykk.tistory.com/82 를 참고하여 작성하였습니다.


두 데이터프라임을 고유 값인 key를 기준으로 병합할 때 사용합니다.

```python
## merge()
# 기준열 이름이 같을 때
pd.merge(left, right, on = '기준열', how = '조인방식')

# 기준열 이름이 다를 때
pd.merge(left, right, left_on = '왼쪽 열', right_on = '오른쪽 열', how = '조인방식')
```



*   left : 왼쪽 데이터프레임

*   right : 오른쪽 데이터프레임

*   on : (두 데이터프레임의 기준열 이름이 같을 때) 기준열

*   how : 조인 방식 {'left', 'right', 'inner', 'outer'} 기본값은 'inner'

*   left_on : 기준열 이름이 다를 때, 왼쪽 기준열

*   right_on : 기준열 이름이 다를 때, 오른쪽 기준열



ㅤ

### **1) how : 결합 방법**

스크린샷 2022-05-07 오후 1.13.47.png

*   left join
  
  왼쪽 데이터프레임을 기준으로 결합합니다. 오른쪽 데이터프레임에 없는 값은 NaN으로 나타냅니다.

  pd.merge(left, right, on = 'col', how = 'left')

```python
# 예시 데이터 생성
fruit = pd.DataFrame({'Num':[123, 456, 789, 1011, 1112], 'Fruit':['Apple', 'Banana', 'Cherry', 'Lemon', 'Peach']})
grade = pd.DataFrame({'Num':[123, 789, 1314], 'Grade':['A', 'B', 'C']})
print(fruit)
print(grade)
```

```python
pd.merge(fruit, grade, on = 'Num', how = 'left')
```

ㅤ

*   Right Join

  오른쪽 데이터프레임을 기준으로 조인합니다. 왼쪽 데이터프레임에 없는 값은 NaN으로 나타냅니다.

  pd.merge(left, right, on = 'col', how = 'right')

```python
pd.merge(fruit, grade, on = 'Num', how = 'right')
```

ㅤ

*  Inner Join

  교집합을 의미합니다. 양쪽에 공통으로 있는 값만 나타냅니다. 

  pd.merge(left, right, on = 'col', how = 'inner')

```python
pd.merge(fruit, grade, on = 'Num', how = 'inner')
```

ㅤ

*  Outer Join

  모든 값이 나타나도록 합니다. 왼쪽 데이터프레임과 오른쪽 데이터프레임에 없는 값들은 NaN으로 나타냅니다.

  pd.merge(left, right, on = 'col', how = 'outer')

```python
pd.merge(fruit, grade, on = 'Num', how = 'outer')
```

ㅤ

## **5. Datetime형 처리**

Datatime는 날짜와 시간을 처리할 수 있는 파이썬 내장 모듈입니다.

Datatime 모듈 내에 있는 클래스 중 많이 쓰는 것 위주로 소개해드리겠습니다!

### **1) timedelta (주, 일, 시, 분, 초, 밀리 초, 마이크로 초)**

datetime 내장 모듈의 **timedelta 클래스**는 **기간을 표현**하기 위해서 사용됩니다.

timedelta 클래스의 생성자는 주, 일, 시, 분, 초, 밀리 초, 마이크로 초를 인자로 받습니다.

<br>

timedelta 객체는 내부적으로 일, 초, 마이크로 초 단위만 저장하기 때문에 위와 같이 해당 정보만 표시됩니다.

```python
# timedelta 클래스 import
from datetime import timedelta

# 기간 표현
timedelta(days=5, hours=17, minutes=30)
```

<br>

파이썬은 timedelta 객체와 함께 **산술/대소 연산자를 사용**할 수 있기 때문에 파이썬의 날짜/시간 계산은 다른 언어에 비해서 **매우 간결하고 직관적**입니다.

timedelta는 today + one_week처럼 더해서 다음주를 구하거나, one_week * 2처럼 곱해서 2주의 기간을 구하거나, one_week < two_weeks처럼 대소 관계를 나타내는 등 자유롭게 산술/대소 연산자를 사용할 수 있습니다.

```python
from datetime import date, timedelta
today = date.today()
print(today)

# 1주의 기간을 담은 변수 생성
one_week = timedelta(weeks=1)
print(one_week)

# 다음주 구하기
next_week = today + one_week
print(next_week)

# 2주의 기간을 담은 변수 생성
two_weeks = one_week * 2
print(two_weeks)

# 두 변수의 대소 관계
print(one_week < two_weeks)

# 내용이 같은지 파악
print(two_weeks == timedelta(weeks=14))

# 저번주 구하기
last_week = next_week - two_weeks
print(last_week)
```

### **2) timezone**

**timezone 클래스**는 **시간대를 표현**하기 위해서 사용됩니다.
<br><br>

timezone 클래스의 생성자는 **UTC 기준으로 시차를 표현**하는 timedelta 객체를 인자로 받아 timezone 객체를 생성해줍니다.

예를 들어, 한국은 UTC 기준으로 9시간이 빠르므로 다음과 같이 timezone 객체를 생성할 수 있습니다.
<br><br>

## **6. 데이터 인코딩 (edcoding)**

사이킷런의 머신러닝 알고리즘은 문자열 값을 입력 값으로 허용하지 않기 때문에 모든 문자열 값을 숫자형으로 변환해야 합니다.

이 과정을 **인코딩(encoding)**이라고 합니다.

문자열 피처에는 카테고리 피처와 텍스트형 피처가 있습니다.

*   **카테고리형 피처**는 카테고리 의미대로 특정 분류에 쓰이는 경우입니다. 

*   **텍스트형 피처**에서 주민번호나 단순 문자열 id와 같은 데이터의 경우 예측 성능을 떨어뜨리는 경우가 많기 때문에 삭제하는 것이 좋습니다.


ㅤ

### **1) 레이블 인코딩(label encoding)**

레이블 인코딩이란 카테고리 피처를 코드형 숫자 값으로 변환하는 과정을 말합니다. 

즉, **문자열 데이터를 숫자형으로 바꾸는 과정**입니다.



ㅤ

사이킷런의 레이블 인코딩은 **LabelEncoder 클래스**로 구현합니다. LabelEncoder를 객체로 생상한 후 **fit(), transform()**을 호출해 레이블 인코딩을 수행합니다.
<br><br>

*  fit(): 문자열 데이터를 숫자형으로 바꾸는 메소드입니다.

*  transform(): 변환된 숫자형 데이터 값을 반환하는 메소드입니다.

```python
from sklearn.preprocessing import LabelEncoder

items=['TV','냉장고','전자렌지','컴퓨터','선풍기','선풍기','믹서','믹서']

# LabelEncoder를 객체로 생성한 후 , fit( ) 과 transform( ) 으로 label 인코딩 수행.
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
print('인코딩 변환값:',labels)
```

ㅤ

위의 경우에는 데이터가 작아 물자열 값이 어떤 숫자 값으로 인코딩 되었는지 직관적으로 알 수 있지만, 데이터가 큰 경우에는 **classes_** 속성값으로 확인할 수 있습니다.

classes_ 속성은 0번부터 순서대로 변환된 인코딩 값에 대한 원본 값을 가지고 있습니다.

```python
print('인코딩 클래스:',encoder.classes_)
```

ㅤ

**inverse_transform()**을 통해 인코딩된 값을 다시 디코딩할 수 있습니다.

```python
print('디코딩 원본 값:',encoder.inverse_transform([4, 5, 2, 0, 1, 1, 3, 3]))
```

<br><br>ㅤ

한편, 레이블 인코딩에는 예측 성능을 떨어뜨리는 문제점이 있습니다.ㅤ

예를 들어, 상품 데이터의 상품 구분이 냉장고, 티비, 전자레인지와 같은 문자열 데이터에서 숫자형으로 변환할 때 이러한 문제점이 발생합니다.
<br><br>

냉장고 => 1

티비 => 2

전자레인지 => 3

<br>

위 예시처럼 냉장고, 티비, 전자레인지에 숫자를 대입할 때

해당 숫자의 크기와는 아무 연관이 없지만 ML 알고리즘에 들어가 해당 모델이 학습을 진행하는 경우, 숫자의 크기에 의해 예측이 잘못될 가능성이 있습니다.
<br><br>

이런 문제점을 해결하기 위해 **원-핫 인코딩(One-Hot encoding)**이 생겼습니다.

<br><br>

### **2) 원-핫 인코딩(One-Hot encoding)**

**원-핫 인코딩**은 피처 값의 유형에 따라 새로운 피처를 추가해 **고유 값에 해당하는 칼럼에만 1**을 표시하고 **나머지 칼럼에는 0**을 표시하는 방식입니다.
<br><br>

즉, 행 형태로 되어 있는 피처의 고유 값을 **열 형태로 차원을 변환**한 뒤 고유 값에 해당하는 칼럼에만 1을 표시하고 나머지 칼럼에는 0을 표시합니다.
<br><br>

**[변환할 때 주의할 점]**

*   OneHotEncoder로 변환하기 전에 모든 문자열 값이 숫자형 값으로 변환되어야 합니다.

*   입력 값으로 2차원 데이터가 필요합니다.

ㅤ

먼저 레이블 인코딩을 통해 문자열 데이터를 숫자형으로 변환합니다.

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

items=['TV','냉장고','전자렌지','컴퓨터','선풍기','선풍기','믹서','믹서']

# 먼저 숫자값으로 변환을 위해 LabelEncoder로 변환합니다.
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
```

ㅤ

image.png

그림 **<숫자로 인코딩 표>**처럼 TV가 0, 냉장고가 1, 믹서가 2, ... 로 상품분류가 숫자로 변환되었습니다.
<br><br>

숫자로 인코딩 된 값을 그림 **<원-핫 인코딩 표>**처럼 바꾸기 위해, reshape(-1, 1)를 이용하여 2차원 데이터로 변환합니다.

```python
# 2차원 데이터로 변환합니다.
labels = labels.reshape(-1,1)
```

ㅤ

마지막으로 원-핫 인코딩 과정을 통해 그림 **<원-핫 인코딩 표>**처럼 만들어줍니다.

```python
# 원-핫 인코딩을 적용합니다.
oh_encoder = OneHotEncoder()
oh_encoder.fit(labels)
oh_labels = oh_encoder.transform(labels)

print('원-핫 인코딩 데이터')
print(oh_labels.toarray())
print('원-핫 인코딩 데이터 차원')
print(oh_labels.shape)
```

8개의 행과 1개의 칼럼을 가진 원본 데이터 => 8개의 행과 6개의 칼럼을 가진 데이터로 변환,
<br><br>

TV : 0

냉장고 : 1

믹서 : 2

선풍기 : 3

전자레인지 : 4

컴퓨터 5

로 각각 첫 번째 칼럼, 두 번째 칼럼, ... ,여섯 번째 칼럼에 인코딩 된 것을 확인할 수 있습니다.
<br><br>

첫 번째 행이 TV이므로 첫 번째 행의 첫 번째 칼럼은 1이고 나머지는 모두 0입니다.

ㅤ

**[get_dummies()]**

*   get_dummies()는 판다스에서 원-핫 인코딩을 더 쉽게 지원하는 API입니다.

*   문자열 카테고리 값을 숫자형으로 변환할 필요없이 바로 변환할 수 있습니다.

```python
import pandas as pd

df = pd.DataFrame({'item':['TV','냉장고','전자렌지','컴퓨터','선풍기','선풍기','믹서','믹서'] })
pd.get_dummies(df)
```

## **7. 스케일링 (Scaling)**

### **1) 스케일링(scaling)이란?**

서로 다른 변수의 **값 범위를 일정한 수준으로 맞춰주는 작업**이며 피처 스케일링(Feature Scaling)이라고도 불립니다.
<br><br>

여기서 피처 스케일링의 피처는 무엇일까요?

피처(Feature)는 **데이터에 담겨있는 값**을 의미합니다.
<br><br>







**[주의할 점]**

*   스케일링하는 과정에서 모든 특성의 범위를 유사하게 만드는 것은 중요하지만, 특성에 따라 각기 다른 스케일링을 적용하는게 유리할 수도 있기 때문에 **모두 같은 분포로 만들 필요는 없습니다.**
<br>

*   scikit-learn 의 scaler를 사용할 때, 

  fit 메소드는 훈련 데이터에만 적용해, 훈련 데이터의 분포를 먼저 학습하고
  그 이후, transform 메소드를 훈련 데이터와 테스트 데이터에 적용해 스케일을 조정해야합니다.

  자세히 말하자면 훈련 데이터로 fit()을 이용해 스케일링 기준을 만든 후 테스트 데이터로 다시 새롭게 스케일링 기준을 만드는 것이 아니라 훈련 데이터로 얻은 fit의 결과로 transform을 진행합니다.
  <br><br>

  즉, **훈련 데이터에는 fit_transform() 메소드**를 적용하고, **테스트 데이터에는 transform() 메소드**를 적용해야 합니다.
<br><br>

 > 이때, fit_transform() 은 fit 과 transform 이 결합된 단축 메서드입니다.

<br>

*   transform( )을 호출할 때 스케일 변환된 데이터 셋이 넘파이의 ndarry로 반환되기 때문에, 이를 **DataFrame으로 변환**해야 합니다.


<br>

**[스케일링(Scaling)의 필요성]**

*   머신러닝 모델이 특정 데이터의 **편향성을 갖는 걸 방지**할 수 있습니다.
<br><br>

*   머신러닝을 위한 데이터셋을 정제할 때, 특성별로 데이터의 스케일이 다르다면 **데이터 간의 상관성이 없다고 판단**할 수 있습니다.

  ex)
  x1는 0과 1 사이의 값을, x2는 1000과 10000000사이의 값을, y는 1000과 10000000사이의 값을 가진다고 가정한다면 
  
  x1의 특징은 y를 예측하는데 큰 영향을 주지 않는다고 생각할 수 있습니다.

### **3) 표준화 (Standardization)**



*   표준화는 특성들의 **평균을 0, 분산을 1**로 만드는 것입니다. 즉, 특성들의 분포를 **정규분포로 만드는 과정**입니다.
<br><br>

**[표준화를 하는 이유]**

머신러닝에서 사용하는 Support Vector Machine, Linear Regression, Logistic Regression 모델은 데이터가 가우시안 분포를 가지고 있다고 가정하여 구현되어 있어서 사전에 학습 데이터에 관해 표준화를 적용하는 것이 모델의 예측 성능 향상에 중요합니다.



<br>

표준화의 수식은 다음과 같습니다.

스크린샷 2022-05-15 오후 5.27.07.png

표준화를 통해 변환될 피처 x의 새로운 i번째 데이터를 xi_new라고 한다면, 이 값은 원래 값에서 피처 x의 평균을 뺀 값을 피처 x의 표준편차로 나눈 값으로 계산할 수 있습니다.

사이킷런에서는 표준화를 쉽게 지원하기 위해 **StandardScaler 클래스**를 제공합니다.

<br><br>

#### **[StandardScaler]**

*  **StandardScaler()**를 사용하여 개별 피처를 평균이 0, 분산이 1인 값으로 변환해주는 표준화 클래스 입니다.

*  최솟값과 최댓값의 크기를 제한하지 않기 때문에, 어떤 알고리즘에서는 문제가 있을 수 있으며 **이상치에 매우 민감**합니다. 이상치 파악에 용이하기 때문에 이상치를 제거할 수 있습니다.

*  회귀보다 **분류에 유용**한 클래스입니다.

<br>

그럼 StandardScaler가 어떻게 데이터 값을 변환하는지 데이터 세트로 확인해 보겠습니다!

```python
from sklearn.datasets import load_iris
import pandas as pd

# 붓꽃 데이터 셋을 로딩하고 DataFrame으로 변환합니다.
iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)

print('feature 들의 평균 값')
print(iris_df.mean())
print('\nfeature 들의 분산 값')
print(iris_df.var())
```

붓꽃 데이터 셋을 이용해 StandardScaler를 하기 위해서는 먼저 데이터 셋을 로딩한 후 DataFrame으로 변환해야 합니다.

그 후, mean(), var() 메소드를 이용하여 붓꽃 데이터 셋의 평균과 분산을 출력합니다.
<br><br>

결과 값을 보면 각 피처들의 평균 값과 분산 값이 달라 스케일링이 필요합니다.


<br>

이제 StandardScaler를 이용하여 각 피처를 한 번에 표준화하여 변환해보겠습니다!

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler객체 생성
scaler = StandardScaler()

# StandardScaler 로 데이터 셋 변환. fit( ) 과 transform( ) 호출.
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)

# transform( )시 scale 변환된 데이터 셋이 numpy ndarry로 반환되어 이를 DataFrame으로 변환
iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)

print('feature 들의 평균 값')
print(iris_df_scaled.mean())
print('\nfeature 들의 분산 값')
print(iris_df_scaled.var())
```

**StandardScaler()**로 StandardScaler 객체를 생성한 후, fit( ) 과 transform( ) 메소드에 변환 대상 피터 데이터 세트를 입력하고 호출하면 간단하게 변환됩니다.

결과 값을 보면 모든 칼럼 값의 평균이 0에 아주 가까운 값으로, 그리고 분산은 1에 아주 가까운 값으로 변환된 것을 확인할 수 있습니다.

<br>

StandardScalar는 각 특성의 평균을 0, 분산을 1로 변경하여 모든 특성이 같은 크기를 가지게 하고 뒤에서 소개할 MinMaxSclaer는 매우 다른 스케일의 범위를 0과 1사이로 변환됩니다.

두 방법은 최댓값과 최솟값 크기를 제한하지 않아 한 가지 문제가 발생합니다.

바로 이상치의 영향을 크게 받는다는 것입니다. 

그것을 해결할 수 있는 스케일링 방법이 있는데, 바로 RobustScaler입니다.
<br><br>

그럼 StandardScaler와 비슷하지만 평균과 분산대신 중앙값과 사분위값을 사용하는 RobustScaler에 대해 소개하겠습니다!

<br>

#### **[RobustScaler]**



*   평균과 분산대신 **중앙값과 사분위값**을 사용하여 **이상치에 영향을 최소화**할 수 있습니다.

*   RobustScaler()를 사용하여 **중앙값이 0**, **IQR(Q3-Q1)이 1(최댓값 1, 최솟값 -1)**이 되도록 변환합니다.

*   모든 피처가 같은 크기를 같는다는 점이 StandardScaler와 유사합니다.



<br>

RobustScaler의 수식은 다음 그림과 같습니다.

데이터값에서 중앙값을 뺀 값을 IQR(Q3-Q1)의 값으로 나눈 값입니다.

스크린샷 2022-05-16 오전 10.59.25.png

<br>

그럼 간단하게 RobustScaler를 구현해보겠습니다.

간단한 예시로 넘파이를 이용하여 0 ~ 10의 숫자가 차례대로 있는 x1 column과 0 ~ 10의 제곱수가 차례로 있는 x2 column 데이터를 생성합니다.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'x1' : np.arange(11), 'x2' : np.arange(11) ** 2})

df
```

<br>

RobustScaler를 import한 후, RobustScaler()를 사용하여 변형 객체를 생성합니다.

앞서 만든 df 데이터를 fit_transform()로 분포를 저장한 후 스케일링을 진행합니다.



```python
from sklearn.preprocessing import RobustScaler

# RobustScaler 객체 생성
scaler = RobustScaler()

# RobustScaler로 데이터 세트 변환
df_robust = scaler.fit_transform(df)

# transform( )시 scale 변환된 데이터 셋이 numpy ndarry로 반환되어 이를 DataFrame으로 변환
df_robust1 = pd.DataFrame(df_robust, columns = ['x1_robust', 'x2_robust'])
df_robust1
```

<br>

```python
print(df_robust1.describe())

Q1 = df_robust1.quantile(.25)
Q2 = df_robust1.quantile(.50)
Q3 = df_robust1.quantile(.75)
IQR = Q3 - Q1

print('median: ', Q2)
print('IQR: ', IQR)
```

describe()를 사용한 결과값에는 중앙값인 50%가 각각 0, IQR인 75%-25%가 각각 '5.000000e-01 - -5.000000e-01'로 1, '0.630000 - -0.370000'로 1이 된 것을 확인할 수 있습니다.
<br><br>

다른 방법으로는 quantile()를 사용하여 직접 Q1, Q2, Q3를 구해 중앙값은 0, IQR값은 1로 변형된 것을 확인할 수 있습니다.

### **4) 정규화 (Normalization)**



*   정규화는 **서로 다른 피처의 크기를 통일**하기 위해 크기를 변환하는 과정입니다. 즉, 개별 데이터의 크기를 모두 **똑같은 단위**로 변경하는 것입니다.

*   일반적으로 **[0, 1]범위의 분포**로 조정합니다.

*   데이터에서 이상치의 존재 유무는 정규화했을 때의 결과에서 큰 차이를 보입니다. 그래서 앞서 언급했듯이 표준화의 결과에서 **이상치를 제거한 후 정규화**를 하는 것이 좋습니다.
<br><br>


**[정규화를 하는 이유]**

*   각 칼럼에 들어있는 데이터의 상대적인 크기로 분석결과가 달라질 수 있기 때문에 이러한 상대적인 크기 차이를 제거할 필요가 있습니다.

*   정규화를 하지 않는다면 **수치가 큰 데이터가 결과에 큰 영향**을 주어 좋은 결과를 얻을 수 없습니다. 또한 **변화패턴이 잡히지 않아** 머신러닝을 진행할 때 **학습이 불가능**합니다.
<br><br>

  ex) A변수는 0 ~ 1000까지의 값을, B변수는 0 ~ 10까지의 값을 가진다면 상대적으로 큰 숫자 값을 갖는 A변수의 영향력이 커지게 됩니다. 그래서 분석을 하기 앞서 상대적인 차이를 제거해야 합니다.

<br>

정규화의 수식은 다음과 같습니다.

원래 데이터 값에서 피처 x의 최솟값을 뺀 값을 피처 x의 최댓값과 최솟값의 차이로 나눈 값으로 계산할 수 있습니다.

스크린샷 2022-05-15 오후 5.27.34.png

그럼 정규화를 쉽게 하도록 지원하는 클래스인 MinMaxScaler에 대해 알아보겠습니다!

<br>

#### **[MinMaxScaler]**



*  **MinMaxScaler()**를 사용하여 각 피처의 최솟값과 최댓값을 기준으로 **[0, 1] 사이 범위**에 균등하게 값을 배정하는 정규화 방법입니다.

*   **음수 값**이 있으면 **[-1, 1] 사이의 값**으로 변환합니다.

*   데이터의 분포가 **정규분포를 따르지 않을 때**, MinMaxScaler를 이용할 수 있습니다.



<br>

그럼 StandardScaler를 구현할 때 사용한 iris 데이터를 이용하여 MinMaxScaler가 어떻게 동작하는지 확인해보겠습니다!

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler객체 생성
scaler = MinMaxScaler()

# MinMaxScaler 로 데이터 셋 변환. fit() 과 transform() 호출.
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)

# transform()시 scale 변환된 데이터 셋이 numpy ndarry로 반환되어 이를 DataFrame으로 변환
iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)

print('feature들의 최소 값')
print(iris_df_scaled.min())

print('\nfeature들의 최대 값')
print(iris_df_scaled.max())
```

MinMaxScaler를 import한 후 MinMaxScaler 객체를 새로 생성합니다.

fit() 과 transform()를 사용하여 iris 데이터를 MinMaxScaler 데이터 셋으로 변환시킵니다.
<br><br>

transform()를 사용할 때, scale 변환된 데이터 셋이 numpy ndarry로 반환되기 때문에 이를 DataFrame으로 변환합니다.

모든 피처들의 최솟값과 최댓값이 각각 0과 1로 스케일링이 적용된 것을 확인할 수 있습니다.

<br>

MinMaxScaler와 유사하지만 다른 모듈이 있는데요.

바로 MaxAbsScaler입니다.

<br>

#### **[MaxAbsScaler]**



*   MaxAbsScaler는 각 특성의 **절대값**이 0과 1 사이가 되도록 변환하는 과정입니다.

*   즉, 모든 값은 **[-1, 1] 사이의 값**으로 표현되며 데이터가 **양수일 경우 MinMaxScaler와 유사**합니다.

*   MinMaxScaler와 마찬가지로 **이상치에 매우 민감**하기 때문에 이상치를 제거하고 MaxAbsScaler를 진행하는 것이 좋습니다.




<br>

RobustScaler를 구현할 때 생성한 df 데이터를 이용하여 MaxAbsScaler가 어떻게 사용되는지 확인해보겠습니다!

```python
df
```

먼저 df에 있는 데이터를 다시 확인해보겠습니다.

df는 x1 column에는 0 ~ 10의 숫자가 차례로, x2 column에는 0 ~ 10의 제곱수가 차례로 있는  데이터입니다.
<br><br>
그럼 MaxAbsScaler로 스케일링을 진행하겠습니다.

<br>

MaxAbsScaler를 import한 후, MaxAbsScaler 객체를 생성합니다.

fit()과 transform()를 사용하여 스케일링을 진행합니다.

앞서 다른 스케일링 방법과 마찬가지로 numpy ndarry로 반환된 데이터 셋을 DataFrame으로 변환합니다.

```python
from sklearn.preprocessing import MaxAbsScaler

# MaxAbsScaler 객체 생성
maxabs_scaler = MaxAbsScaler()

# MaxAbsScaler로 데이터 세트 변환, 분포 저장
maxabs_scaler.fit(df)
maxabs_scaled = maxabs_scaler.transform(df)

# transform( )시 scale 변환된 데이터 셋이 numpy ndarry로 반환되어 이를 DataFrame으로 변환
df_maxabs = pd.DataFrame(maxabs_scaled, columns = ['x1_maxabs', 'x2_maxabs'])
df_maxabs
```

앞서 소개한 4가지 스케일러는 각 **특성(열)의 통계치를 이용**하여 진행됩니다.

이외에도 위의 방식들과 다르게 적용되는 방식있습니다.

바로 **Normalizer**로, 일반적으로 데이터 전처리 과정에서 잘 사용되지 않아 간단하게 소개해보겠습니다


<br>

#### **[Normalizer]**

사이킷런에서 제공하는 **Normalizer** 모듈은 일반적인 정규화와 약간의 차이가 있습니다.
<br><br>

Normalizer 모듈은 **선형대수에서의 정규화 개념이 적용**되었으며, **개별 벡터의 크기를 맞추기 위해 변환**하는 것을 의미합니다.

즉, 개별 벡터를 모든 피처 벡터의 크기로 나눠 줍니다.

<br>

Normalizer의 수식은 다음과 같습니다.

세 개의 피처 x, y, z가 있다고 하면 새로운 데이터 xi_new는 원래 값에서 세 개의 피처의 i번째 피터 값에 해당하는 크기를 합한 값으로 나눠 줍니다.

<br>


**<특징>**

*   Normalizer는 각 **샘플(행)마다 적용**되는 방식입니다.

*   **한 행의 모든 특성들 사이의 유클리드 거리(두 점 사이의 최단거리)가 1**이 되도록 스케일링을 진행합니다. 

*   Normalizer는 일반적인 데이터 전처리에서 사용되는 것이 아니라 모델, 특히 **딥러닝 내 학습 벡터**에 적용합니다.

*   특히 피처들이 다른 단위(키, 나이, 소득 등)라면 더욱 사용하지 않습니다.
