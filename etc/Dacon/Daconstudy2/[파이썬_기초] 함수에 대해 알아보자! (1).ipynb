**사각형 넓이 구하기 함수**

```python
# 사각형 면적
def rectangle(width, height):
    area = width * height
    return area

rectangle(10, 20)


area = rectangle(10, 20)   #변수로 받아 출력 가능
print(area)


x = 10     #함수 안에 변수를 넣을 있음
y = 20
rectangle(x, y)


rectangle(11.2, 2)  #실수를 넣을 수 있음


def rectangle(width, height):  
    
    return width * height     #마찬가지로 바로 return으로 출력가능

rectangle(x, y)
```

**사용자 정의 함수 또다른 예시(합 구하는 함수)**

```python
#사용자 정의 함수 생성
def f(x):
    
    return 2 * x ** 2 + 4 * x + 2

f(3)

#for문에서 사용
data = [2.3, 1.7, 9.1]
for x in data:
    print(x, f(x))


#합 구하는 함수
sum(data)    #내장함수


def my_sum(data):  #사용자 정의 함수
    
    sum_ = 0
    for x in data:
        sum_ += x

    return sum_

my_sum(data)
```

**함수와 함수의 계산**

함수와 함수의 연산이 가능합니다.

```python
#함수와 함수의 계산
def g(x):
    
    return x ** 2

g(2.4)

f(x) + g(x)

f( g(x) )    #합성함수 계산
g( f(x) )

f(g(x)) + g(f(x))
```

**작은 수 구하는 함수**

작은 수를 출력하는 함수인 min()을 사용자 정의 함수로 만들어 봅시다.

```python
#작은 수 구하는 함수
min(2, 3)
min(3, 3)

def my_min(a, b):
    
    if a <= b:     #함수 안에 if문 포함 가능
        return a
    else:
        return b

my_min(2, 3)
```

**평균 & 분산을 구하는 함수**

numpy 라이브러리의 평균을 구하는 함수 mean(), 분산을 구하는 함수 var()을 사용자 정의 함수로 만들어 봅시다.

```python
#평균 구하는 함수
def my_mean(data):
    
    return my_sum(data) / len(data)   

my_mean(data)


#분산 구하는 함수
def my_var(data):
    
    mx = my_mean(data)
    
    sum_ = 0
    for x in data:
        sum_ += (x - mx) ** 2
        
    return sum_ / (len(data) - 1)  #연산의 결과 리턴 가능

my_var(data)


#함수의 연산 결과 리턴하는 함수
def describe(data):
    
    return my_mean(data), my_var(data) #두 개 리턴 가능 

describe(data)

#함수 반환값을 언팩킹해서 받기
mean, var = describe(data)   #반환값을 각각 mean, var변수에 넣기
print(mean, var)


#위의 식은 my_mean(), my_var()에서 둘다 평균을 계산
#계산 효율성 더 높이는 함수 만들기
#표본평균, 표본분산, 표본표준편차 구하는 함
def describe(data):
    
    df = len(data) - 1   #자유도
    mx = my_mean(data)   #표본평균
    
    sum_ = 0
    for x in data:
        sum_ += (x - mx) ** 2
        
    mv = sum_ / df    #표본분산
    
    mstd = mv ** 0.5  #표본표준편차
    
    return mx, mstd, mv

mean, std, var = describe(data)

print(mean, std, var)
 
```
