# **문자열**


*  문자열은 시퀀스 자료형이며 숫자형과 자료저장 방식이 다릅니다. 

*  또한, 문자열은 저장공간을 많이 차지합니다.

ㅤ

 ***컴퓨터의 자료 저장방식***


*   기본적으로 컴퓨터는 0 or 1만 이해하고 사용가능 => 이진수로 주저장공간에 저장

*   2개의 이진수 값 저장 => 1비트

*   1바이트 = 8비트 // 2**8 = 256, 8개의 비트로 256개의 조합을 가질 수 있습니다.


ㅤ
*   **예시**

    sys 모듈의 getsizeof를 사용하여 문자를 얼마만큼의 크기로 저장하는지 확인해봅시다.



    sys 모듈 : 플랫폼(하드웨어, 소프트웨어)에 가지고 있는 시스템 정보에 접근할 수 O






```python
# 저장공간 크기: 바이트
import sys
print(sys.getsizeof("a"))
print(sys.getsizeof("ab"))
print(sys.getsizeof("abc"))

print(sys.getsizeof("가"))
```

# **문자표**

*   문자 => 숫자로 인식, 그 숫자들이 갖는 순서를 정리해놓은 표를 문자표(unicode)라고 합니다.
*   unicode는 지구상에 존재하여 실제 사용하는 모든 언어의 문자들을 다 가지고 있습니다.







```python
# 문자표
print(ord('A'))  #순서 확인
print(ord('a'))

print(chr(65))  #문자 확인 
```

# **문자열을 리스트로**



*   파이썬에서 대표적으로 시퀀스 자료형을 저장할 수 있는 객체 ==> 리스트
*   문자열은 리스트 중 특수한 경우입니다.






```python
#리스트로 문자열을 리스트와 똑같이 다룰 수 O
# 인덱싱
a = 'abcde'
print(a[0], a[4])
print(a[-1], a[-5])


# 슬라이싱
print(a[1:])
print(a[2:-1])

print(a[::2])
print(a[::-1])


b = 'ㄱㄴㄷㄹㅁ'  #한글도 가능

print(a + b)  #리스트 연산

print(b * 5)

for c in b:
    print(c, ord(c))
    
print(ord('가'))
print(ord('힣')) 
    
ord('힣') - ord('가') #한글의 발음가능한 수
```

# **문자열 함수**



*   문자열이 가지고 있는 함수를 이용하여 전처리를 할 수 있다.

 즉, 문자열에 포함되어 있는 것 중  분석의 대상이 아닌 문자열을 정리한다.




ㅤ



*   upper : 대문자로 변환
*   lower : 소문자로 변환

   **영어를 자연어처리할 때 소문자, 대문자가 단어의 의미 이해에 영향 X => 소문자로 변환하는 과정 O**
ㅤ
ㅤ
ㅤ
ㅤ

ㅤ
*   title : 제목은 첫번째가 대문자이므로 그렇게 변환
*   capitalize  => 문장의 첫글자는 대문자이므로 그렇게 변환
ㅤ

ㅤ

*   count : 해당 문자가 문자열 내에 몇번 포함되어 있는지 횟수 체크
*   find : 해당 문자가 문자열 내에 몇번째에 위치해 있는지 체크
*   rfind : r => reverse의 약자, 뒤에서 부터 몇번째 위치에 있는지 체크


ㅤ
*   startswith : 해당 문자로 시작하는지 확인
*   endswith : 해당 문자로 끝나는지 확인


ㅤ
*   strip : 문자열 내에 있는 공백을 제거

   **자연어 처리에서 불필요한 빈칸, 띄어쓰기는 노이즈라고 불림, 제거 필요**


ㅤ
*   rstrip : 오른쪽에 있는 공백만 제거
*   lstrip : 왼쪽에 있는 공백만 제거


ㅤ
*   isdigit : 해당 문자가 숫자인지 아닌지 판별















```python
# 문자열 함수

#upper => 대문자로 변환
a = 'abcde'
print(len(a))
print(a.upper())


#lower => 소문자로 변환
#영어를 자연어처리할 때 소문자, 대문자가 단어의 의미 이해에 영향 X => 소문자로 변환하는 과정 O
a = 'ABCDE'
print(a.lower())


#title => 제목으로 첫번째를 대문자
a = 'hello world!'
print(a.title())


#capitalize  => 문장으로 첫번째를 대문자
a = 'hello world!'
print(a.capitalize())


#count => 몇번 포함되어 있는지 횟수 체크
print(a.count('l'))


#find => 몇번째에 위치해 있는지 체크
print(a.find('l'))
print(a.find('lo'))


#rfind => r(reverse), 뒤에부터 몇번째 위치에 있는지 체크
print(a.rfind('l'))


#startswith  => 해당 문자로 시작하는지 확인
print(a.startswith('h'))


#endswith => 해당 문자로 끝나는지 확인
print(a.endswith('w'))


#strip => 공백 제거
#자연어 처리에서 불필요한 빈칸, 띄어쓰기는 노이즈라고 불림, 제거 필요
a = '  가나         '
ord(' ')
print(a.strip())

a = '  가나         '
print(a.rstrip())  #오른쪽 공백 제거
print(a.lstrip())  #왼쪽 공백 제거


#isdigit => 해당 문자가 숫자인지 아닌지 판별
print('3'.isdigit())
```

# **특수문자**

```python
# 특수 문자: \' \" \b \t, \n \e
print('a \t b')  #tab
print('a \n b')  #enter
print('a \b b')  #backspace


a = "It\'s Ok." # \' => '표현
print(a)

# """   """ : 세개의 문장을 하나의 문자열로 인식 
a = """
It's OK.
I'm Happy.
See you."""

print(a)
```
