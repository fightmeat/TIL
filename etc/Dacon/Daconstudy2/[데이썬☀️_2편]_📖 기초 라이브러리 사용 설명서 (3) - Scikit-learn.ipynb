````python
import numpy as np
from sklearn import preprocessing
````

````python
# ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ ÏûÖÎ†•
input_labels = ['red','black','red','green']
````

````python
#Î†àÏù¥Î∏î Ïù∏ÏΩîÎçî ÏÉùÏÑ±ÌõÑ ÏïûÏóêÏÑú Ï†ïÏùòÌïú Î†àÏù¥Î∏îÎ°ú ÌïôÏäµÏãúÌÇ§Í∏∞.
encoder = preprocessing.LabelEncoder()
encoder.fit(input_labels)
labels = encoder.transform(input_labels)
````

>LabelEncoder(): Î†àÏù¥Î∏î Ïù∏ÏΩîÎî©ÏùÑ Íµ¨ÌòÑÌïòÎäî ÌÅ¥ÎûòÏä§
