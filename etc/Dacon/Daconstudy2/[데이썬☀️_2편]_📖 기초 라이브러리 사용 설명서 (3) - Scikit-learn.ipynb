#### **인코딩**

- preprocessing 모듈을 이용
- 문자열을 **숫자형 코드값**으로 처리

<br>

**[레이블 인코딩]**   

- 문자열 값을 숫자로 변환하는 인코딩 방식
- 각각의 문자열 값에 숫자를 1개씩 대치

````python
import numpy as np
from sklearn import preprocessing
````

````python
# 샘플 데이터 입력
input_labels = ['red','black','red','green']
````

````python
#레이블 인코더 생성후 앞에서 정의한 레이블로 학습시키기.
encoder = preprocessing.LabelEncoder()
encoder.fit(input_labels)
labels = encoder.transform(input_labels)
````

>LabelEncoder(): 레이블 인코딩을 구현하는 클래스   
>Fit(): 인코딩을 학습하는 메소드 -> 실행 후 문자열을 숫자값으로 학습  
>Transform(): 학습한 숫자값을 실제로 변환


이때, 숫자로 변환 된 인코딩 값은 숫자 값의 크고 작음에 대한 특성이 작용하여
단순히 숫자를 표시로 인식하는 것이 아니라 숫자나 중요도로 인식 될 수 있습니다.

이를 해결해기 위한 방법으로 다음 인코딩 방법이 있습니다.

<br>

**[원-핫 인코딩(one-hot encoding)]**
- 레이블 인코딩의 문제를 해결 할 수 있는 인코딩 방법

- 2차원 구현으로 레이블 인코딩을 커쳐야 함
- 고유값에 해당하는 칼럼에만 1을 표시하고 나머지는 0으로 표시하는 방식

아래 그림을 참조하면 두 인코딩 방법간의 **차이**를 더 명확하게 확인 할 수 있습니다.


인코딩 그림.JPG

이제 앞서 레이블 인코딩으로 2차원 변환 되어진 데이터를 **원-핫 인코딩**으로 변환 시켜 보겠습니다.

````python
# 레이블 인코딩 2차원 변환
labels = labels.reshape(-1,1)
````

</br>변환된 데이터를 **원-핫 인코딩** 적용합니다.

````python
#원-핫 인코딩
encoder2 = preprocessing.OneHotEncoder()
encoder2.fit(labels)
labels2 = encoder2.transform(labels)
labels2
````

#### **[train_test_split]**

 데이터를 **train data/ test data**로 나누는 방법입니다.




  </br></br>
데이터 분리 실습은 scikit-learn의 내장 데이터 세트인 iris를 이용하여 진행합니다.  

실습을 위하여 scikit-learn의 내장 데이터 세트인 iris를 불러옵니다.


````python
from sklearn.datasets import load_iris
iris = load_iris()
````

</br> 불러온 데이터 세트의 **피처와 레이블 값**을 각각의 변수에 저장합니다.

````python
# 피처만으로 된 데이터 저장 
iris_data=iris.data

# 레이블(결정 값) 데이터 저장 
iris_label=iris.target

iris_label
iris.target_names
````

</br> model_selection의 train_test_split 모듈을 불러옵니다.


````python
from sklearn.model_selection import train_test_split
````

</br> 데이터를 train과 test 데이터로 **분할** 합니다.

````python
X_train,X_test,y_train,y_test = train_test_split(iris_data,iris_label,test_size=0.2,random_state=11)
X_train
X_test
````


```
- test_size : 학습/테스트 데이터 세트의 비율
- random_state : 지정하지 않으면 매번 다른 데이터 세트 만들어짐


-X_train, X_test : 모의고사 데이터를 학습/테스트 데이터로 나눔
-y_train, y_test  : 정답 데이터를 학습/테스트 데이터로 나눔
```  




>**주요 파라미터**    
>
>**Test_size**: 전체 데이터에서 테스트 데이터 세트 크기를 알마로 샘플링할 것인가를
	 결정. 디폴트 값은 0.25  	 
>**Train_size**: 전체 데이터에서 학습용 데이터 세트 크기를 얼마로 샘플링할 것인가를
	  결정. 통상적으로 test_size 파라미터를 이용하기 때문에 잘 사용하지 않음  
>**Shufle**: 데이터를 분리하기 전에 데이터를 미리 섞을지 여부 결정. 디폴트 값은 true.
          사용시 데이터를 분산시켜 좀 더 효율적인 학습을 할 수 있도록 도와줌.  
>**Random_state** : 호출 할 때마다 동일한 학습/테스트용 세트를 생성하기 위하여
	          주어지는 난수 값. Train_test_spilt는 호출 시 무작위로 데이터를
		호출하여 random_state 값을 지정하지 않으면 실행 시마다 다른
	          데이터 세트를 만들어 냄


데이터 세트를 분리 할 때 **샘플 수가 충분하지 못하다면** 데이터를 나누는 방식에 따라 **정확도의 차이**가 크게 날 수 있습니다.   </br></br>

이를 방지하기 위하여 여러번의 검증을 하는 다음의 방법들이 있습니다.

<br>

#### **[K-fold 방식]**

 - Training data를 여러 개의 묶음으로 나누어 모의고사를 실행하는 것
</br></br>

- 순서

  ① Fold는 각각 하나의 학습 과정이고, Fold의 숫자(k)만큼 학습/검증의 과정을 
    거침

  ➁ 각 Fold 마다 training data 는 k-1개, test data는 1개로 생성

  ➂ 각 fold 결과의 평균이 최종 학습 결과가 됨</br></br>

❗ 그렇다면 lable data가 불균형할 때는 어떻게 해야 할까요? 


<br>

#### **[stratified KFold]**

- lable data가 **불균형**할 때 사용하는 방식

- fold를 나눌 때 원본 데이터의 lable 분포를 고려한 뒤 학습/테스트 데이터 분리 시 **원본 데이터 비율과 동일하게** 분배

#### **[Cross_val_score]**

- Sklearn.model_selection 모듈에 내장되어있는 메소드

- 교차 검증을 보다 간편하게 해주는 API

- Kfold로 데이터를 학습하고 예측하려면   

   1) 폴드 세트 설정  

   2) for loop 반복 학습 및 테스트 데이터의 인덱스 추출  

   3) 반복적인 학습과 예측 수행 후 성능 반환 과정을 거쳐야 한다

  **Cross_val_score()** 는 이런 일련의 과정을 한꺼번에 수행해 주는 API

- 폴드 세트 분할 시 **stratifiedKFold**를 사용
