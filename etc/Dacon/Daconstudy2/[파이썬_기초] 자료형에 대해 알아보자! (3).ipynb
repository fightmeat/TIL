# **dictionary**




*   사전처럼 **단어 : 단어 뜻**, 이렇게 짝을 이루는 자료형을 사용할 때 

*   **{key: value, key: value, ...}** 의 형태로 형성함


ㅤ







```python
# dictionary
# {key: value, key: value, ...}

dic = {'a': 101, 'b': 202, 'c': 303}

print(dic)

len(dic)  #길이보단 몇개의 아이템을 가지고 있냐고 부름

print(dic['a'])

dic['a'] = 1001
print(dic)
```

***keys() / values() / items()***

*   keys() : key만 출력하는 것
*   values() : value만 출력하는 것

*   items() : item을 튜플의 형태로 리스트로 출력





```python
# keys() / values() / items()

print(dic.keys())
print(dic.values())
print(dic.items())   #튜플의 형태로 리스트로 출력


for k, v in dic.items():
    print('Key: %s, Value : %s' % (k, v))


dic_value = dict()
for k, v in dic.items():
    dic_value[v] = k     #key => value, value => key로 바꿈
    
print(dic_value)   
```

# **deque**



*   파이썬 내장되어 있는 collections 패키지 내에 있는 모듈입니다.

*   stack & cue 에 관한 모든 기능을 제공합니다.


*   리스트와 같은 형식일 때 사용가능합니다.




ㅤ
*   append : 맨 오른쪽에 새로운 값을 넣음

*   appendleft : 맨 왼쪽에 새로운 값을 넣음

*   pop : 맨 끝에 있는 값을 제거

*   popleft : 맨 왼쪽에 있는 값을 제거

*   rotate : 인덱스 번호를 바꿈







```python
# deque

from collections import deque

# append / pop / appendleft / rotate ....

deque_list = deque()
for i in range(5):
    deque_list.append(i)
    
print(deque_list)

deque_list.append(5)
print(deque_list)

deque_list.appendleft(-1)
print(deque_list)

deque_list.pop()
print(deque_list)

deque_list.rotate(3)
print(deque_list)
```

# **Counter**

리스트와 같은 시퀀스 자료형을 딕셔너리 자료형으로 변환 가능합니다.

```python
from collections import Counter

text = list('Hello World!')
cnt = Counter(text)
print(cnt)
```

# **namedtuple**



*   객체를 생성하듯이 tuple 생성 가능
*   딕셔너리의 키처럼 직관적으로 원소에 접근 가능



```python
# namedtuple

from collections import namedtuple

Point = namedtuple('Point', ['x', 'y'])  #생성

p = Point(x = 11, y = 22)  #값 넣기

print(p)
print(p.x, p.y)

p[0]
p[1]
```
