# **리스트 연산**

리스트의 연산으로는 덧셈(결합)과 곱셈(반복)이 있음

ㅤ



*   **덧셈**

    리스트 자체를 더하는 것 => 리스트 안에 있는 요소끼리 덧셈 X



```python
#기본 리스트 생성
one = [1, 1.645, 9 / 5]
two = ['하나','one', 1.0]
```

```python
#덧셈
three = one + two
print(len(three))
print(three)
```

*  **곱셈**

   리스트 자체를 반복 => 리스트 안에 있는 요소끼리 곱셈(반복) X

  [1, 2, 3] * 2ㅤ=> [1, 2, 3, 1, 2, 3] (O)   
ㅤㅤㅤㅤㅤ    ㅤ => [1, 1, 2, 2, 3, 3] (X)        

```python
#곱셈
print(one * 3)  #3번 반복
print('#' * 5)  #문자열도 리스트처럼 인덱싱, 슬라이싱 가능
```

*   **in 연산**

    리스트 안에 해당 요소가 들어있는지 True or False로 나타남



```python
#in
print(1 in one)
print('one' in two)   #문자열도 확인 가능
print((9 / 5) in one)
print('하나' in one)
print(4 in one)
```

*   **추가 사항**

    문자열도 리스트처럼 사용가능



```python
#문자열을 리스트처럼 사용가능
name = 'joy'
print(name[-1])
print(name[1:3])

print(name[::-1])

print('y' in name)
```

# **요소 추가**



*   **list()**

    새로운 리스트를 ㅤ**'생성'**ㅤ하는 함수



```python
#list
color = list()   # color = []
print(len(color))
```

*   **append(값)**

    리스트 안에 ㅤ**'요소'**ㅤ 넣는 함수

```python
# append
color.append('red')
print(color)
color.append('green')
print(color)
color.append('blue')
print(color)
print(len(color))
```

*   **extend([요소1, 요소2, ...])**

    리스트를 **추가**할 수 있는 함수, 한번에 여러 개의 요소를 추가할 때 사용

    **<append()와 차이점>**

    : extend()는 **리스트의 형태** 즉, **x.extend([요소1, 요소2, ...])** 이렇게 추가
      
      append()는 **요소와 리스트** 둘 다 가능 

      ㅤbut append() 리스트의 형태로 추가 
      
     ㅤ  => 리스트 안에 리스트 [요소1, 요소2, ..., [추가1, 추가2]] 형태로 추가 



```python
#extend
color.extend(['white', 'black'])
print(color)

color.append(['gray', 'dodgerblue']) #리스트안에 리스트를 넣을 수 있음
print(color)
```

*   **insert(위치, 값)**

    **지정한 위치**에 **삽입**하는 함수



```python
#insert
color.insert(0, 'orange') #지정한 위치에 삽입
print(color)
color.insert(3, 'purple')
print(color)
```

# **요소 제거**





*   **remove(요소)**

    
ㅤㅤㅤ리스트 속 **요소 제거**하는 함수

ㅤㅤㅤ**<주의 사항>**

ㅤㅤㅤ: 같은 요소가 여러 개일 경우, 앞에 있는 것만 제거



```python
# remove
color.remove('purple')
print(color)
```

*   **del(위치)**

ㅤㅤㅤ**원하는 위치**(인덱스)에 있는 요소 제거할 수 있는 함수



```python
#del
del color[0]   #0번째 요소 없애고 싶을 때
print(color)

del color[1:3] #1부터 2번째 있는 요소 없애고 싶을 때
print(color)
```

ㅤㅤㅤ**원하는 위치**(인덱스)에 있는 요소 제거할 수 있는 함수



```python
#pop
color.pop(0)
print(color)
```
