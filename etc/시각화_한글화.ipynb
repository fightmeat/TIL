### 그리고 런타임을 눌러서 다시시작을 하고나서 밑에 코드를 실행해야한다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family='NanumGothic')
```

```python
plt.title('한글 제목')
plt.plot([10,20,30,40],[1,4,9,16])
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.show()
```

### 한글화에 - 를 출력하려면

```python
plt.rcParams["axes.unicode_minus"]=False
```

```python
plt.title('한글 제목')
plt.plot([-10,-20,-30,-40],[1,4,9,16])
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.show()
```
