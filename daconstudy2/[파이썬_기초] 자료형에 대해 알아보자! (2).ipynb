# **stack**



*   입력받은 순서대로 저장공간 내에 쌓는다.
*   **Last In First Out(LIFO)** : 젤 마지막에 입력된(저장된) 값을 젤 처음에 이용한다. 



***stack은 push와 pop을 이용합니다.***

push를 이용하여 차곡차곡 쌓고 pop을 이용하여 맨 마지막에 입력한 값을 가지고 계산에 이용

*   **push**

   하나의 값을 추가하는 것을 말하며, 리스트에 append로 추가하는 것과 같습니다.
*   **pop**

   마지막에 입력된 인덱스의 값을 출력합니다.
 


```python
# push : 리스트에 append가 일어난 것, 하나의 값을 추가
a = list()
a.append('a')
a.append('b')
print(a)

# pop : 마지막에 입력된 인덱스 값을 출력
a.pop()


#문자열로 받은 것을 형변환(캐스팅)을 통해 하나하나씩 push가 일어나 리스트가 생성
word = input("Input a word: ")
word_list = list(word)
print(word_list)


#pop이 일어나 뒤에서부터 출력
for _ in range(len(word_list)):
    print(word_list.pop())
```

# **cue**



*   **First In First Out(FIFO)** : 맨 처음에 입력된 값을 맨 처음에 이용합니다.



```python
# cue : First In First Out(FIFO)

fifo = [1, 2, 3, 4, 5]

for _ in range(len(fifo)):
    print(fifo.pop(0))
```

# **tuple**


*   리스트와 다른 점 : 인덱스의 값 바꾸는 것 허용 X , 리스트와 다르게 **"()"를 사용**

*   원자료가 계산 과정에서 값의 변화를 불가능하게 하기 위해 고정하기 용이함



```python
#tuple 
tup = (1, 2, 3, 4, 5)

for t in tup:
    print(t)
    
print(tup[0])

tup[0] = 10  #값 넣지 X

a, b, c, d, e = tup  #언팩킹
print(a)

#연산 가능
print(tup + tup)
print(tup * 2)
```

# **set**

#    **집합 연산**
 
   
*   **합집합 : or**
 
   union, | 을 사용하여 표현


*   **교집합 : and**

    intersection, &을 사용하여 표현



*   **차집합**

    difference, -을 사용하여 표현








```python
# 집합 연산
A = {1, 2, 3, 4, 5}
B = {3, 4, 5, 6, 7}

# 합집합 : or
A.union(B)
A | B

# 교집합 : and
A.intersection(B)
A & B


#차집합
A.difference(B)
B.difference(A)

A.difference(B) | B.difference(A)  #교집합을 제외한

A - B
B - A

A - B | B - A
```
