# **비교 연산자**



*   **x < y,ㅤ x > y**

ㅤㅤx가 y보다 **작은지 큰지** 비교하는 연산자


```python
# x < y, x > y
x = 3
y = 10

x < y

x > y
```

*    **x <= y, ㅤx >= y**

ㅤㅤㅤx가 y보다 **작거나 같은지** 또는 **크거나 같은지** 비교하는 연산자


```python
#x <= y, x >= y
x <= y

x >= y
```

*   **x == y,ㅤ x != y**

ㅤㅤㅤ두 변수의 값이 서로 **같은지, 다른지 비교**하는 연산자




```python
#값이 서로 같은지 비교
x == y
x != y
```

ㅤ



*   **x ㅤisㅤ y,ㅤ ㅤx ㅤis notㅤ y**

ㅤㅤㅤ개체가 서로 **같은, 다른 개체**인지 비교하는 연산자



```python
# is : 개체가 서로 같은 개체인지 비교

x = [1, 2, 3]
y = [1, 2, 3]

x == y

x is y  # x와 y는 서로 다른 메모리 공간에 저장되어 있음


z = x
z is x  # x와 z는 서로 같은 메모리 공간에 저장되어 있음

z == x
```

#   **논리 연산자**





*   **and**

ㅤㅤㅤ두 값이 **모두 True 일때** True**(교집합)**



```python
# and : 두 값이 모두 True 일때 True(교집합)
(7 > 5) and (10 > 5)
(7 < 5) and (10 > 5)
(7 < 5) and (10 < 5)
```

*   **or**

ㅤㅤㅤ두 값 중 **하나만 True 일 경우** True**(합집합)**




```python
# or : 두 값 중 하나만 True 일 경우 True(합집합)
(7 > 5) or (10 < 5)
(7 < 5) or (10 < 5)
```

*   **not**

```python
# not : 논리역
not(7 < 5)
not(7 > 5)
```

# **조건문**





*   **한 줄 if문**



```python
x,y = 10, 6

#one line if statement

x if x < y else y            #x < y이면 x값, 아니면 y값

result = x if x < y else y

print(result)
```

*   **if- else**

ㅤㅤㅤif에 있는 조건문이 참 => if문 ㅤ**"ㅤ:ㅤ"**ㅤ밑줄에 있는 수행할 문장을 실행

ㅤㅤㅤif에 있는 조건문이 거짓 => else문 ㅤ**"ㅤ:ㅤ"**ㅤ밑줄에 있는 수행할 문장을 실행


```python
#if- else

if x < y:     # <, > 비교연산자
    print(x)
else:
    print(y)
```

*   **ifㅤ -ㅤ elifㅤ -ㅤ else**

ㅤㅤif에 있는 조건문이 참 => if문  ㅤ**"ㅤ:ㅤ"**ㅤ밑줄에 있는 수행할 문장을 실행

ㅤㅤif에 있는 조건문이 거짓 => elif에 있는 조건문이 참 => elif문 **"ㅤ:ㅤ"**밑줄에 있는 수행할 문장을 실행ㅤㅤㅤ


ㅤㅤㅤㅤㅤㅤㅤㅤㅤㅤㅤㅤㅤ=> elif에 있는 조건문이 거짓 => else문 **"ㅤ:ㅤ"**밑줄에 있는 수행할 문장을 실행ㅤㅤㅤ

```python
# if - elif - else

score = 55

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
    

print(score < 90 and score >= 80)

80 <= score < 90     # 파이썬에서만 가능
```
