## **1. Seaborn이란?** 


matplotlib을 기반으로 하는 python 데이터 시각화 라이브러리입니다. 

matplotlib을 기반으로 다양한 색상 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지입니다. 
<br><br>
함수를 그려보기에 앞서 seaborn 함수에서 사용하는 주요 파라미터들을 알아보도록 하겠습니다.
<br><br>
**[ parameters ]**

* **x, y**ㅤ:ㅤx, y축의 위치를 지정하는 변수
* **hue** ㅤ:ㅤ다른 색상의 선을 생성하는 그룹화 변수
* **size**ㅤ:ㅤ너비가 다른 선을 생성하는 그룹화 변수 (마커의 경우, 마커의 크기가 다르게 나타남)
* **style**ㅤ:ㅤ다른 스타일의 마커 또는 선을 생성하는 그룹화 변수
* **data**ㅤ:ㅤ데이터 구조 입력
* **palette**ㅤ:ㅤhue 체계를 매핑할 때 사용할 색상 선택

ㅤ

## **3. Relational plots**

relational plot이란 x값과 y값의 관계를 그려주는 관계형 그래프를 의미합니다. 

<br>

### **1) lineplot**



lineplot()는 선 그래프를 그리는 함수로, 데이터가 시간 또는 순서에 따라 어떻게 변화하는지 보여주기 위해 사용합니다.

```python
april_flights = flights.query("month == 'Apr'")
sns.lineplot(data=april_flights, x="year", y="passengers")
plt.show()
```

* flights 데이터에서 month 값이 "Apr"인 데이터만 april_flights에 저장했습니다.

* april_flights 데이터로 x값이 "year"이고 y값이 "passengers"인 선 그래프를 그렸습니다.

* seaborn 함수는 show() 함수를 사용하지 않아도 그래프가 플롯팅되지만, 그림이 표시되지 않는 오류를 방지하기 위해 실행해주었습니다. 

ㅤ

```python
sns.lineplot(data=flights, x="year", y="passengers")
plt.show()
```

* **data=flights**

  4월 데이터로만 그렸던 lineplot과는 달리 전체 데이터셋을 할당하면 passengers에 대한 평균과 95% 신뢰구간을 표시합니다.

ㅤ

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.show()
```

* **hue="month"**

  hue 파라미터에 "month"를 대입한 결과, "month"의 값에 따른 별도의 선이 다른 색상으로 나타났습니다.

ㅤ

```python
sns.lineplot(data=flights, x="year", y="passengers", style="month")
plt.show()
```

* **style="month"**

  style 파라미터에 "month"를 대입한 결과, "month"의 값에 따른 별도의 선이 다른 선 스타일로 나타났습니다.

ㅤ

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

ㅤ


* **hue="day"**

* **style="time"**

  style 파라미터에 "time"을 대입해 "time"의 값에 따라 점의 스타일이 다르게 나타났습니다.


* **hue="size", size="size"**

  hue와 size 파라미터 모두 "size"를 대입해 "size"의 값이 커짐에 따라 색상이 점점 어두워지고, 점의 크기가 점점 커지는 것을 확인할 수 있습니다.

## **4. Distribution plots**

distribution plot이란 변수 값의 분포를 나타내기 위한 그래프를 의미합니다.

분포를 나타내는 가장 대표적인 그래프인 **히스토그램**을 그리는 두 가지 함수에 대해 알아보도록 하겠습니다. 

히스토그램은 도수분포표를 그래프로 나타낸 것으로, 가로축은 계급 세로축은 도수입니다.

<br>

### **1) displot**



displot()는 히스토그램을 그리는 함수입니다.

```python
sns.displot(data=penguins, x="flipper_length_mm")
plt.show()
```

* penguins 데이터로 x값이 "flipper_length_mm"이고 값인 히스토그램을 그렸습니다.


ㅤ

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="kde")
plt.show()
```

* **kind="kde"**

  kind 파라미터에 "kde"를 대입하면 커널 밀도 추정을 사용하여 일변량 또는 이변량 분포를 플롯팅합니다.

* 이를 사용하면, 다중 분포를 그릴 때 덜 복잡하고 해석하기 쉬운 플롯을 생성할 수 있습니다.


ㅤ

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="ecdf")
plt.show()
```

* **kind="ecdf"**

  kind 파라미터에 "ecdf"를 대입하면 경험적 누적 분포 함수를 플롯팅합니다.

* 경험적 누적 분포 함수는 각 고유 값에 대한 관찰의 비율 또는 개수를 나타냅니다.

ㅤ

```python
sns.displot(data=penguins, x="flipper_length_mm", kde=True)
plt.show()
```

* **kde=True**

  kde 파라미터에 True를 대입하면 kde 곡선이 추가된 히스토그램을 그릴 수 있습니다.

ㅤ

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", kind="kde")
plt.show()
```

*     **hue="species", kind="kde"**

  hue 파라미터에 "species", kind 파라미터에 "kde"를 대입하면, 각 species의 kde 플롯을 다른 색상으로 하여 별도로 그릴 수 있습니다.

ㅤ

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()
```

* **hue="species", multiple="stack"**

  hue 파라미터에 "species", multiple 파라미터에 "stack"을 대입하면, 각 species의 히스토그램을 다른 색상으로 하여 하나의 막대에 여러 개의 막대를 쌓는 형식으로 그릴 수 있습니다.

ㅤ

### **2) histplot**



histplot()는 히스토그램을 그리는 함수입니다.

```python
sns.histplot(data=penguins, x="flipper_length_mm")
plt.show()
```

* penguins 데이터로 x값이 "flipper_length_mm"인 히스토그램을 그렸습니다.

ㅤ

```python
sns.histplot(data=penguins, y="flipper_length_mm")
plt.show()
```

* **y="flipper_length_mm"**

  "flipper_length_mm"을 y축에 할당하면, 수평 방향의 히스토그램을 그릴 수 있습니다.

ㅤ

```python
sns.displot(data=penguins, x="flipper_length_mm", binwidth=7)
plt.show()
```

* **binwidth=7**

  binwidth 파라미터를 이용하면 히스토그램 막대의 너비를 지정할 수 있습니다.

ㅤ

```python
sns.displot(data=penguins, x="flipper_length_mm", bins=20)
plt.show()
```

* **bins=20**

  bins 파라미터를 이용하면 히스토그램 막대의 개수를 지정할 수 있습니다.

ㅤ

```python
sns.histplot(data=penguins, x="flipper_length_mm", kde=True)
plt.show()
```

* **kde=True**

  kde 파라미터에 True를 대입하면, kde 곡선을 히스토그램 위에 추가할 수 있습니다.

ㅤ

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()
```

* **hue="species", multiple="stack"**

  hue 파라미터에 "species", multiple 파라미터에 "stack"을 대입하면, 각 species의 히스토그램을 다른 색상으로 하여 하나의 막대에 여러 개의 막대를 쌓는 형식으로 그릴 수 있습니다


ㅤ

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", element="step")
plt.show()
```

* **element="step"**

  element 파라미터에 "step"을 대입하면, 겹치는 막대의 시각적인 어려움을 해결할 수 있습니다.

ㅤ

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", element="poly")
plt.show()
```

* **element="poly"**

  element 파라미터에 "poly"를 대입하면, 각 막대의 중앙에 꼭짓점이 있는 다각형의 형태인 히스토그램을 그릴 수 있습니다. 

* 이 그래프는 분포의 모양을 더 쉽게 파악할 수 있지만, 히스토그램이라는 사실이 명확하지 않다는 단점이 있습니다.

## **5. Categorical plots**

categorical plot은 범주형 변수와 연속형 변수간의 관계를 나타내기 위한 그래프를 의미합니다.
<br><br>

**[ 스타일 지정 ]**

categorical plot을 그리기에 앞서 plot의 스타일, 팔레트, 글꼴, 글꼴 크기 등을 설정하는 함수인 set에 대해 알아보려고 합니다.

**set_theme()**은 plot에 대한 시각적 테마를 설정하는 함수입니다.
<br><br>

ex)

**sns.set_theme()**

→ 기본 테마가 활성화 됩니다.

**sns.set_theme(style="whitegrid", palette="pastel")**

→ 스타일이 "whitegrid"이고, 색상이 "pastel"인 테마가 활성화 됩니다.

ㅤ

### **1) barplot**



barplot()은 막대 그래프를 그리는 함수로, 직사각형 막대로 범주형 데이터를 요약합니다. 막대 그림은 숫자 변수의 중심 경향 추정치를 나타내며, 막대 상단의 검은 선은 해당 추정치 주변의 불확실성을 나타내는 오차 막대입니다.

```python
sns.set_theme(style="whitegrid")     # 스타일 지정
sns.barplot(data=tips, x='day', y='total_bill')
plt.show()
```

* **sns.set_theme(style="whitegrid")**

  plot의 스타일을 "whitegrid"로 설정했습니다.

* tips 데이터로 x값이 범주형 변수인 "day"이고 y값이 숫자형 변수인 "total_bill"인 수직 막대 그래프를 그렸습니다.

ㅤ

```python
sns.barplot(data=tips, x="total_bill", y="day")
plt.show()
```

* **x="total_bill", y="day"**

  x와 y값을 반대로 할당하면 수평 막대 그래프를 그릴 수 있습니다.

ㅤ

```python
sns.barplot( data=tips, x='day', y='total_bill', hue="sex")
plt.show()
```

* **hue="sex"**

  hue 파라미터에 "sex"를 대입하여 "sex"의 값에 따라 막대의 색상이 다르게 나타났습니다.

ㅤ

```python
sns.barplot(data=tips, x='day', y='total_bill', palette="Blues_d")
plt.show()
```

* **palette="Blues_d"**

  palette 파라미터를 이용해 막대의 색상을 "Blues_d"로 지정했습니다.

ㅤ


### **2) boxplot**




boxplot()은 범주에 대한 분포를 표시하는 상자 그림을 그릴 수 있는 함수입니다.

```python
sns.boxplot(x = tips["total_bill"])
plt.show()
```

* **x=tips["total_bill"]**

  x값에 칼럼명 대신 <u>데이터["칼럼명"]</u> 형식으로 입력할 수도 있습니다.

ㅤ

```python
sns.boxplot(data=tips, x="day", y="total_bill")
plt.show()
```

* **x="day", y="total_bill"**

  x값에 범주형 변수 "day"를 y값에 숫자형 변수 "total_bill"을 대입하여 "day"로 그룹화된 수직 상자 그림을 그렸습니다.

ㅤ

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker",  palette="Set3")
plt.show()
```

* **x="day", hue="smoker"**

  x값에 범주형 변수 "day"를 hue 파라미터에 범주형 변수 "smoker"를 대입하여 "day"와 "smoker"로 중첩 그룹화된 상자 그림을 그렸습니다.

* **palette="Set3"**

  palette 파라미터를 이용하여 상자 그림의 색상을 "Set3"으로 설정했습니다.

ㅤ


### **3) countplot**



countplot()은 막대를 사용하여 범주형 변수의 관측값 개수를 표시하는 함수입니다.

```python
sns.set_theme(style="darkgrid")     # 스타일 지정
sns.countplot(data=tips, x="day")
plt.show()
```

* **sns.set_theme(style="darkgird")**

  plot의 스타일을 "darkgrid"로 설정했습니다.

* x값에 범주형 변수 "day"를 대입하여 "day"의 관측값의 개수를 표시하는 countplot을 생성했습니다.

ㅤ

```python
sns.countplot(data=tips, x="day", hue="time")
plt.show()
```

* **hue="time"**

  hue 파라미터에 "time"을 대입하여 두 개의 범주형 변수 "day"와 "time"에 대한 관측값의 개수를 표시하는 그래프를 그렸습니다.

ㅤ

```python
sns.countplot(data=tips, y="day", hue="time")
plt.show()
```

* **y="day"**

  y값에 범주형 변수 "day"를 대입하면 countplot의 막대를 가로로 플롯팅할 수 있습니다.

ㅤ

```python
sns.countplot(data=tips, x="day", palette="Set3")
plt.show()
```

* **palette="Set3"** 

  palette 파라미터를 이용해 막대의 색상을 "Set3"로 설정했습니다.

ㅤ


### **4) stripplot**



stripplot()은 하나의 변수가 범주형인 산점도를 그리는 함수로 범주형 변수에 들어있는 각 범주별 데이터의 분포를 확인할 수 있습니다. 데이터의 값이 중복되는 경우 데이터 포인트가 중복되게 그려집니다.

```python
sns.set_theme(style="whitegrid")
sns.stripplot(x=tips["total_bill"])
plt.show()
```

* **sns.set_theme(style="whitegrid")**
  
  plot의 스타일을 "whitegrid"로 설정했습니다.
  
* x값에 "total_bill"값을 대입하여 단일 수평 스트립 플롯을 생성했습니다.

ㅤ

```python
sns.stripplot(data=tips, x="day", y="total_bill")
plt.show()
```

* **x="day", y="total_bill"**

  x값에 범주형 변수 "day"를 y값에 숫자형 변수 "total_bill"을 대입하여 범주형 변수 "day"로 스트립을 그룹화한 그래프를 그렸습니다.

ㅤ

```python
sns.stripplot(data=tips, x="total_bill", y="day")
plt.show()
```

* **x="total_bill", y="day"**

  x와 y값을 반대로 대입하면 스트립을 가로로 그릴 수 있습니다.

ㅤ

```python
sns.stripplot(data=tips, x="sex", y="total_bill", hue="day")
plt.show()
```

* **hue="day"**

  hue 파라미터에 범주형 변수 "day"를 대입하여 "day"별로 점이 색상이 다르게 나타났습니다.


ㅤ


### **5) swarmplot**



swarmplot()은 점이 겹치지 않는 범주형 산점도를 그리는 함수입니다. stripplot과 유사하지만, 데이터의 분산까지 고려하여 점이 겹치지 않도록 조정됩니다.

```python
sns.swarmplot(x=tips["total_bill"])
plt.show()
```

* x값에 tips 데이터의 "total_bill"을 대입하여 단일 수평 swarm plot을 그렸습니다.

ㅤ

```python
sns.swarmplot(data=tips, x="day", y="total_bill")
plt.show()
```

* **x="day"**

  x값에 범주형 변수 "day"를 대입하여 "day"로 그룹화된 수직 swarm plot을 그렸습니다.

ㅤ

```python
sns.swarmplot(data=tips, x="total_bill", y="day")
plt.show()
```

* **x="total_bill", y="day"**

  x와 y값을 반대로 대입하면 수평으로 swarm plot을 그릴 수 있습니다.

ㅤ

```python
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex")
plt.show()
```

* **hue="sex"**

  hue 파라미터에 "sex"를 대입하여 두 번째 범주형 변수 "sex"를 다른 색의 점으로 나타냅니다.

ㅤ
