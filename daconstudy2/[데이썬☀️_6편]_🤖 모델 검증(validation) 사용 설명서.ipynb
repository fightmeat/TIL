## **1. 모델 검증의 필요성**

### **1) 모델 검증이란?**

**모델 검증**이란 말 그대로 모델을 설계한 후 모델을 검증하는 단계를 말합니다.
<br>

**모델 검증이 필요한 이유**는 모델을 적합하게 잘 설계했는지, 즉 **예측**을 잘하는지 알기 위해서 입니다.

<br>

### **2) 표본내 성능 검증 & 표본외 성능 검증(교차검증)**

기본적으로 **훈련 데이터**를 기반으로 **모델링**하고, **테스트 데이터**로 해당 **모델의 성능**을 측정합니다.

* **표본내 성능 검증** : 훈련 데이터의 종속 변수값을 잘 예측하는지 보는 과정
* **표본외 성능(교차검증)** : 테스트 데이터의 종속 변수값을 잘 예측하는지 보는 과정
<br>

**👉  표본외 성능 검증(교차검증)**은 **예측을 잘하는지** 알기 위해서는 **학습에 사용하지 않는 표본**을 활용해 모델이 잘 돌아가는지 확인합니다.<br>


**👨🏼‍⚕ 표본외 성능 검증(교차검증)을 하는 이유**는 <br>
표본내 성능이 좋으면 표본외 성능이 상대적으로 많이 떨어지는 **과최적화 상태**가 나타날 수 있습니다. <br> 
이런 경우 새로운 데이터를 주었을 때 전혀 예측하지 못해, 예측을 할 수 있는 모형을 만들기 위해 과최적화를 방지하는 **정규화** 과정이 필요합니다.

<br>

## **2. Train test split**

### **1) Train test split이란?**

**교차검증**을 하기 위해 두 종류의 데이터 집합이 필요합니다.
* **Training data set** : 교차검증을 하려면 모델 학습을 위한 데이터 집합
* **Test data set** : 모델 성능 검증을 위한 데이터 집합 <br>


👉 **Train data**에는 예측해야 하는 변수가 포함되어 있지만 **test data**에는 포함되어 있지 않는 것이 특징입니다. <br>
두 데이터 집합 모두 종속 변수값이 있어야 해서 보통 가지고 있는 데이터 집합으로 나누어 사용하며, <br>
**Train test split(학습/검증 데이터 분리) 메소드**를 사용해 학습용 데이터와 검증용 데이터로 **분리**합니다.


<br>

## **3. 교차검증**

### **1) 교차검증이란?**

1.jpg

**교차검증**은 모델을 평가하는 방법 중 하나로, 보통 데이터 셋을 그림과 같이 나누게 됩니다. <br>

Valid data를 **모델의 과적합을 방지**하기 위해 사용되는데, <br>
Train data에서 파생된 데이터로 Train data로 모델을 학습시키는 과정에 Valid data로 중간중간 학습된 모델을 평가합니다. <br>
🧐 만약 Valid data로 검증을 했을 때 평가지표가 좋지 않은 수치로 나타난다면 **과적합을 의심**해야 합니다. <br>

그림과 같은 데이터 분리 후, 모델 검증을 하면 **모델 예측에 있어 과적합을 방지**할 수 있습니다. <br>
보통 Train : Test 데이터 분리 비율은 8:2 ~ 7:3 사이 입니다.



<br>

### **2) K-fold 교차검증이란?**

2.jpg

일반적으로 많이 사용되는 교차검증 방법인 **K-fold 교차검증**에 대해 설명해보겠습니다. <br>

👉 **K-fold 교차검증**은 Train data set 전체를 K개의 학습세트와 검증세트로 나누어, <br>
K만큼의 폴트세트에 K번의 모델 학습과 검증를 수행합니다. 
<br><br>
그림과 같이 **K번의 교차검증 성능을 평균**내어 최종 교차검증 성능으로 계산합니다. <br>
K-fold의 경우 **데이터의 수가 적은 경우** 사용하기 좋은 방법입니다.


<br>

### **3) K-fold 교차검증의 장단점**

다음으로 교차검증의 장단점에 대해 설명해드리도록 하겠습니다. <br><br>

**👍 교차검증의 장점**
1. 모든 데이터 셋을 평가에 반영할 수 있어 데이터의 편중을 막을 수 있음
2. 더욱더 일반적인 모델을 생성 가능하여 과적합을 방지
3. 반복적인 교차검증을 통해 모델의 정확도를 향상 시킬 수 있음
4. 데이터가 부족하여 train과 test로 나누기에 아까운 경우에 활용하기 좋음
<br>

**👎 교차검증의 단점**
1.  k번의 모델을 생성해야 하므로, 모델 생성 및 평가하는 과정에서 많이 시간이 소요됨


<br>

## **3. 교차검증**

### **1) Metrics란?**

3.jpg

Scikit-learn의 metrics 서브패키지에는 **Confusion matrix**라는 훈련된 모델의 성능을 측정하기 위한 **Matrix**가 있습니다. <br><br>

그림은 **Confusion matrix(이진 분류결과표)**입니다. <br>
* **TP(True Positive)**: True를 True로 잘 예측한 것
* **TN(True Negative)**: False를 False로 잘 예측한 것
* **FP(False Positive)** : False를 True로 잘 못 예측한 것
* **FN(False Negative)** : True를 False로 잘 못 예측한 것
<br>

👉  **Confusion matrix**를 활용해 정확도, 정밀도, 재현율 등을 측정할 수 있습니다. 
<br>

그림에서 positive와 negative는 **예측값이 긍정이냐 부정이냐**를 의미하고, <br>
true/false는 **예측값과 실제값이 같은지 다른지**를 나타냅니다.


<br>

### **2) 평가점수**

Scikit-learn의 metrics 서브패키지에서는 예측 성능을 평가하기 위한 다양한 함수를 제공합니다. <br>
그 중 **회귀분석에 유용한 함수**는 다음과 같습니다.
<br>

* **R2_score** : 결정계수
* **Mean_squared_error** : 평균 제곱 오차
* **Median_absolute_error** : 절대 오차 중앙값

<br>

그외에도 많이 사용하여 알아두면 좋은 다양한 평가점수에 대해 설명해보겠습니다.

<br>

#### **📐 정확도(accuracy)**


image.png

**정확도**는 **전체 샘플 중 맞게 예측한 샘플 수의 비율**로, 정확도가 높을 수록 좋은 모형입니다. <br>
일반적으로 학습에 **최적화 목적함수**로 사용됩니다.

<br>

#### **🧪 정밀도(precision)**

image.png

**정밀도**는 **positive 클래스에 속한다고 예측한 샘플 중 실제로도 positive 클래스에 속하는 샘플 수의 비율**로, <br>
이것 또한 높을 수록 좋은 모형입니다.

<br>

#### **🔍 재현율(recall)**

image.png

**재현율**은 **실제 positive 클래스에 속한 표본 중에 positive 클래스에 속한다고 예측한 표본의 수의 비율**로, 
<br> 이것도 높을 수록 좋은 모형입니다. 또한 재현율을 **TPR**(true positive rate) or **민감도**라고 말하기도 합니다.


<br>

#### **💥 위양성율(fall-out)**

image.png

**위양성율**은 **실제 negative에 속하지 않는 표본 중에 negative에 속한다고 출력한 표본의 비율**로, <br>
**FPR**(false positive rate) or 1에서 위양성율의 값을 뺀 값을 **특이도**라고 합니다.

<br>
