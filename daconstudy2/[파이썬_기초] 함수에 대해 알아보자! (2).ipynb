# **함수 안에 함수(로컬 함수)**

for loop 안에 for loop를 넣을 수 있는 만큼 함수 안에도 함수를 넣을 수 있습니다.

```python
 #함수 내에 로컬 함수
def my_mean(data):
    
    def my_sum(data):
        sum_ = 0
        for x in data:
            sum_ += x
        return sum_
    
    return my_sum(data) / len(data)

my_mean(data)
```

# **재귀 함수**

함수 안에 자기 자신을 재호출하는 함수입니다.

```python
# recursion
# recursive function(재귀함수) (함수 안에 자기 함수를 재호출)

def h(x):
    
    return x + 2

h(1)

h( h(1) )   

h( h( h(1) ) )
```

*   **재귀함수의 예시1(팩토리얼)**

n까지의 곱을 계산하는 함수 생성



```python
#재귀함수의 예시1 (팩토리얼)

# 1! = 1
# 2! = (1!) * 2
# 3! = 1 * 2 * 3
# 3! = 2! * 3
# 3! = (1! * 2) * 3
# 3! = ((1!) * 2) * 3
# n! = (n-1)! * n

def factorial(n):
    
    if n == 1:
        return n
    else:
        return factorial(n - 1) * n
    
factorial(3)
```

*  **재귀함수의 예시2**

n까지의 합을 계산하는 함수 생성



```python
#재귀함수의 예시2(n까지의 합계산)
# 1 + 2 + 3
# (1 + 2) + 3
# ((1) + 2) + 3
# sum(n) = sum(n-1) + n

def recursive_sum(n):
    if n == 1:
        return n
    else:
        return recursive_sum(n - 1) + n
    
recursive_sum(10)
```

*   **디폴트 인수**

값을 안넣어도 **고정값으로 계산**, 하지만 **값을 넣으면 넣은 값**으로 계산됩니다. 

그래서 값을 넣지 않은 이상 특정한 값으로 고정되길 원할 때 사용합니다.



```python
#디폴트 인수
def print_value(x, y = 10): # y는 고정값으로 설정 =>디폴트 인수
    
    return x, y

print_value(20)  #y는 10으로 고정값, x의 값만 지정


print_value(20, 1234)  #y가 디폴트 인수지만 값을 넣으면 그걸로 계산
```

*   **가변인수**

**여러가지 인수**를 받고 싶을 때 사용하는데 *args **굵은 텍스트**의 형태로 사용됩니다.

튜플의 형태로 받기 때문에 인덱스의 형태로 값에 접근 가능합니다.



```python
#가변 인수
def variable(a, b, *args): # *args : 가변인수
    return a, b, args  #여기에는 '*'를 넣지 않음

variable(1, 2, 3)

variable(1, 2, 3, 4, 5)  #3, 4, 5 => 팩킹 => (3, 4, 5)


def variable(a, b, *args):
    x, y, z = a, b, args  #언팩킹
    return x, y, z      

variable(1, 2, 3, 4, 5)


def variable(a, b, *args):
    x, y, *z = a, b, args   # *z로 하면 ()로 팩킹이 아니라 리스트 안에 팩킹된다.
    return x, y, z

variable(1, 2, 3, 4, 5)
```

*   **키워드 가변인수**

가변함수와 마찬가지로 여러 개의 인수를 받고 싶을 때 사용하는데 **kwargs의 형태로 사용한다.

딕셔너리(Dictionary)의 형태로 받습니다.

ㅤ

**+딕셔너리**: **Key, Value 한 쌍**으로 대응관계를 나타내는 자료형




```python
#키워드 & 디폴트 인수 사용 
def key_word_variable_length(x = 1, y = 1, z = 3):
    
    return x, y, z

key_word_variable_length()  #디폴트 인수 = 값 입력 안해도 O


#키워드 가변 인수
def key_word_variable_length(a, b, **kwargs):
    
    return a, b, kwargs

key_word_variable_length(1, 2, first = 100, second = 200) #딕셔너리 형태로 받음
```
