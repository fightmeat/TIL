timedeltaëŠ” today + one_weekì²˜ëŸ¼ ë”í•´ì„œ ë‹¤ìŒì£¼ë¥¼ êµ¬í•˜ê±°ë‚˜, one_week * 2ì²˜ëŸ¼ ê³±í•´ì„œ 2ì£¼ì˜ ê¸°ê°„ì„ êµ¬í•˜ê±°ë‚˜, one_week < two_weeksì²˜ëŸ¼ ëŒ€ì†Œ ê´€ê³„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë“± ììœ ë¡­ê²Œ ì‚°ìˆ /ëŒ€ì†Œ ì—°ì‚°ìë¥¼ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

#### **[StandardScaler]**

*  **StandardScaler()**ë¥¼ ì‚¬ìš©í•˜ì—¬ ê°œë³„ í”¼ì²˜ë¥¼ í‰ê· ì´ 0, ë¶„ì‚°ì´ 1ì¸ ê°’ìœ¼ë¡œ ë³€í™˜í•´ì£¼ëŠ” í‘œì¤€í™” í´ë˜ìŠ¤ ì…ë‹ˆë‹¤.

*  ìµœì†Ÿê°’ê³¼ ìµœëŒ“ê°’ì˜ í¬ê¸°ë¥¼ ì œí•œí•˜ì§€ ì•Šê¸° ë•Œë¬¸ì—, ì–´ë–¤ ì•Œê³ ë¦¬ì¦˜ì—ì„œëŠ” ë¬¸ì œê°€ ìˆì„ ìˆ˜ ìˆìœ¼ë©° **ì´ìƒì¹˜ì— ë§¤ìš° ë¯¼ê°**í•©ë‹ˆë‹¤. ì´ìƒì¹˜ íŒŒì•…ì— ìš©ì´í•˜ê¸° ë•Œë¬¸ì— ì´ìƒì¹˜ë¥¼ ì œê±°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

*  íšŒê·€ë³´ë‹¤ **ë¶„ë¥˜ì— ìœ ìš©**í•œ í´ë˜ìŠ¤ì…ë‹ˆë‹¤.

ê·¸ëŸ¼ StandardScalerê°€ ì–´ë–»ê²Œ ë°ì´í„° ê°’ì„ ë³€í™˜í•˜ëŠ”ì§€ ë°ì´í„° ì„¸íŠ¸ë¡œ í™•ì¸í•´ ë³´ê² ìŠµë‹ˆë‹¤!

```python
from sklearn.datasets import load_iris
import pandas as pd

# ë¶“ê½ƒ ë°ì´í„° ì…‹ì„ ë¡œë”©í•˜ê³  DataFrameìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)

print('feature ë“¤ì˜ í‰ê·  ê°’')
print(iris_df.mean())
print('\nfeature ë“¤ì˜ ë¶„ì‚° ê°’')
print(iris_df.var())
```

ë¶“ê½ƒ ë°ì´í„° ì…‹ì„ ì´ìš©í•´ StandardScalerë¥¼ í•˜ê¸° ìœ„í•´ì„œëŠ” ë¨¼ì € ë°ì´í„° ì…‹ì„ ë¡œë”©í•œ í›„ DataFrameìœ¼ë¡œ ë³€í™˜í•´ì•¼ í•©ë‹ˆë‹¤.

ê·¸ í›„, mean(), var() ë©”ì†Œë“œë¥¼ ì´ìš©í•˜ì—¬ ë¶“ê½ƒ ë°ì´í„° ì…‹ì˜ í‰ê· ê³¼ ë¶„ì‚°ì„ ì¶œë ¥í•©ë‹ˆë‹¤.

ê²°ê³¼ ê°’ì„ ë³´ë©´ ê° í”¼ì²˜ë“¤ì˜ í‰ê·  ê°’ê³¼ ë¶„ì‚° ê°’ì´ ë‹¬ë¼ ìŠ¤ì¼€ì¼ë§ì´ í•„ìš”í•©ë‹ˆë‹¤.

ì´ì œ StandardScalerë¥¼ ì´ìš©í•˜ì—¬ ê° í”¼ì²˜ë¥¼ í•œ ë²ˆì— í‘œì¤€í™”í•˜ì—¬ ë³€í™˜í•´ë³´ê² ìŠµë‹ˆë‹¤!

```python
from sklearn.preprocessing import StandardScaler

# StandardScalerê°ì²´ ìƒì„±
scaler = StandardScaler()

# StandardScaler ë¡œ ë°ì´í„° ì…‹ ë³€í™˜. fit( ) ê³¼ transform( ) í˜¸ì¶œ.
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)

# transform( )ì‹œ scale ë³€í™˜ëœ ë°ì´í„° ì…‹ì´ numpy ndarryë¡œ ë°˜í™˜ë˜ì–´ ì´ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜
iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)

print('feature ë“¤ì˜ í‰ê·  ê°’')
print(iris_df_scaled.mean())
print('\nfeature ë“¤ì˜ ë¶„ì‚° ê°’')
print(iris_df_scaled.var())
```

**StandardScaler()**ë¡œ StandardScaler ê°ì²´ë¥¼ ìƒì„±í•œ í›„, fit( ) ê³¼ transform( ) ë©”ì†Œë“œì— ë³€í™˜ ëŒ€ìƒ í”¼í„° ë°ì´í„° ì„¸íŠ¸ë¥¼ ì…ë ¥í•˜ê³  í˜¸ì¶œí•˜ë©´ ê°„ë‹¨í•˜ê²Œ ë³€í™˜ë©ë‹ˆë‹¤.

ê²°ê³¼ ê°’ì„ ë³´ë©´ ëª¨ë“  ì¹¼ëŸ¼ ê°’ì˜ í‰ê· ì´ 0ì— ì•„ì£¼ ê°€ê¹Œìš´ ê°’ìœ¼ë¡œ, ê·¸ë¦¬ê³  ë¶„ì‚°ì€ 1ì— ì•„ì£¼ ê°€ê¹Œìš´ ê°’ìœ¼ë¡œ ë³€í™˜ëœ ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

*   RobustScaler()ë¥¼ ì‚¬ìš©í•˜ì—¬ **ì¤‘ì•™ê°’ì´ 0**, **IQR(Q3-Q1)ì´ 1(ìµœëŒ“ê°’ 1, ìµœì†Ÿê°’ -1)**ì´ ë˜ë„ë¡ ë³€í™˜í•©ë‹ˆë‹¤.

*   ëª¨ë“  í”¼ì²˜ê°€ ê°™ì€ í¬ê¸°ë¥¼ ê°™ëŠ”ë‹¤ëŠ” ì ì´ StandardScalerì™€ ìœ ì‚¬í•©ë‹ˆë‹¤.

ê·¸ëŸ¼ ê°„ë‹¨í•˜ê²Œ RobustScalerë¥¼ êµ¬í˜„í•´ë³´ê² ìŠµë‹ˆë‹¤.

ê°„ë‹¨í•œ ì˜ˆì‹œë¡œ ë„˜íŒŒì´ë¥¼ ì´ìš©í•˜ì—¬ 0 ~ 10ì˜ ìˆ«ìê°€ ì°¨ë¡€ëŒ€ë¡œ ìˆëŠ” x1 columnê³¼ 0 ~ 10ì˜ ì œê³±ìˆ˜ê°€ ì°¨ë¡€ë¡œ ìˆëŠ” x2 column ë°ì´í„°ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'x1' : np.arange(11), 'x2' : np.arange(11) ** 2})

df
```

<br>

RobustScalerë¥¼ importí•œ í›„, RobustScaler()ë¥¼ ì‚¬ìš©í•˜ì—¬ ë³€í˜• ê°ì²´ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.

ì•ì„œ ë§Œë“  df ë°ì´í„°ë¥¼ fit_transform()ë¡œ ë¶„í¬ë¥¼ ì €ì¥í•œ í›„ ìŠ¤ì¼€ì¼ë§ì„ ì§„í–‰í•©ë‹ˆë‹¤.



```python
from sklearn.preprocessing import RobustScaler

# RobustScaler ê°ì²´ ìƒì„±
scaler = RobustScaler()

# RobustScalerë¡œ ë°ì´í„° ì„¸íŠ¸ ë³€í™˜
df_robust = scaler.fit_transform(df)

# transform( )ì‹œ scale ë³€í™˜ëœ ë°ì´í„° ì…‹ì´ numpy ndarryë¡œ ë°˜í™˜ë˜ì–´ ì´ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜
df_robust1 = pd.DataFrame(df_robust, columns = ['x1_robust', 'x2_robust'])
df_robust1
```

<br>

```python
print(df_robust1.describe())

Q1 = df_robust1.quantile(.25)
Q2 = df_robust1.quantile(.50)
Q3 = df_robust1.quantile(.75)
IQR = Q3 - Q1

print('median: ', Q2)
print('IQR: ', IQR)
```

describe()ë¥¼ ì‚¬ìš©í•œ ê²°ê³¼ê°’ì—ëŠ” ì¤‘ì•™ê°’ì¸ 50%ê°€ ê°ê° 0, IQRì¸ 75%-25%ê°€ ê°ê° '5.000000e-01 - -5.000000e-01'ë¡œ 1, '0.630000 - -0.370000'ë¡œ 1ì´ ëœ ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

*   **ìŒìˆ˜ ê°’**ì´ ìˆìœ¼ë©´ **[-1, 1] ì‚¬ì´ì˜ ê°’**ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
