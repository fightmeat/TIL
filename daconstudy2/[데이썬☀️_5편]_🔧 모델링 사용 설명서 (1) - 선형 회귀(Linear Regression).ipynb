```python
# ë ˆì´ë¸” ì¸ì½”ë”©ì„ í•˜ê¸° ìœ„í•œ dictionary map ìƒì„± í•¨ìˆ˜
def make_label_map(dataframe):
    label_maps = {}
    for col in dataframe.columns:
        if dataframe[col].dtype=='object': 
            label_map = {'unknown':0} 
            for i, key in enumerate(train[col].unique()):
                label_map[key] = i+1
            label_maps[col] = label_map
    return label_maps

# ê° ë²”ì£¼í˜• ë³€ìˆ˜ì— ì¸ì½”ë”© ê°’ì„ ë¶€ì—¬í•˜ëŠ” í•¨ìˆ˜
def label_encoder(dataframe, label_map):
    for col in dataframe.columns:
        if dataframe[col].dtype=='object':
            dataframe[col] = dataframe[col].map(label_map[col])
            dataframe[col] = dataframe[col].fillna(label_map[col]['unknown']) #í˜¹ì‹œ ëª¨ë¥¼ ê²°ì¸¡ê°’ì€ unknownì˜ ê°’ìœ¼ë¡œ ì±„ì›Œì¤ë‹ˆë‹¤.
    return dataframe

# ì „ì²˜ë¦¬ ì™„ë£Œí•œ train ë°ì´í„° : train_preprocessed
label_map = make_label_map(train_preprocessed) # train_preprocessed ì‚¬ìš©í•´ label map ìƒì„±
labeled_train = label_encoder(train_preprocessed, label_map) # train_preprocessed ë¼ë²¨ ì¸ì½”ë”©
```

## **2. ê·œì œ ì„ í˜• ëª¨ë¸**

ì´ì „ê¹Œì§€ëŠ” ì„ í˜• ëª¨ë¸ì€ ì‹¤ì œê°’ê³¼ ì˜ˆì¸¡ê°’ì˜ ì°¨ì´ë¥¼ ìµœì†Œí™”í•˜ëŠ” ê²ƒë§Œ ê³ ë ¤í–ˆìŠµë‹ˆë‹¤. ê·¸ëŸ¬ë‹¤ ë³´ë‹ˆ í•™ìŠµ ë°ì´í„°ì— ì§€ë‚˜ì¹˜ê²Œ ë§ì¶°ì§€ê²Œ ë˜ê³ , íšŒê·€ ê³„ìˆ˜ê°€ ì‰½ê²Œ ì»¤ì ¸ ì˜ˆì¸¡ ì„±ëŠ¥ì´ ì‰½ê²Œ ì €í•˜ë˜ëŠ” ë¬¸ì œê°€ ë°œìƒí•©ë‹ˆë‹¤.

### **1) ë¦¿ì§€ íšŒê·€ (Ridge)**

Wì˜ ì œê³±ì— ëŒ€í•´ í˜ë„í‹°ë¥¼ ë¶€ì—¬í•˜ëŠ” L2 ê·œì œë¥¼ ì„ í˜• íšŒê·€ì— ì ìš©í•œ ê²ƒì´ **ë¦¿ì§€(Ridge)** íšŒê·€ì…ë‹ˆë‹¤.

ì‚¬ì´í‚·ëŸ°ì€ **Ridge** í´ë˜ìŠ¤ë¥¼ í†µí•´ ë¦¿ì§€ íšŒê·€ë¥¼ êµ¬í˜„í•˜ê³ , Ridge í´ë˜ìŠ¤ì˜ ì£¼ìš” ìƒì„± íŒŒë¼ë¯¸í„° alphaëŠ” ë¦¿ì§€ íšŒê·€ì˜ alpha L2 ê·œì œ ê³„ìˆ˜ì— í•´ë‹¹í•©ë‹ˆë‹¤. 

alpha ê°’ì„ ì¦ê°€ì‹œí‚¤ë©´ íšŒê·€ ê³„ìˆ˜ê°€ 0ì— ê°€ê¹Œì›Œì§‘ë‹ˆë‹¤.

<br>

í­ê·„ ëª¸ë¬´ê²Œ ì˜ˆì¸¡ ì˜ˆì œë¥¼ Ridge í´ë˜ìŠ¤ë¥¼ ì´ìš©í•´ ë‹¤ì‹œ ì˜ˆì¸¡í•´ë³´ê³ , ì˜ˆì¸¡ ì„±ëŠ¥ì„ cross_val_score()ë¡œ í‰ê°€í•´ë³´ê² ìŠµë‹ˆë‹¤.

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

# alpha=10
ridge = Ridge(alpha=10)
neg_mse_scores = cross_val_score(ridge, feature, target, scoring="neg_mean_squared_error", cv=5)
rmse_scores = np.sqrt(-1 * neg_mse_scores)
avg_rmse = np.mean(rmse_scores)
```

### **2) ë¼ì˜ íšŒê·€ (Lasso)**

Wì˜ ì ˆëŒ“ê°’ì— í˜ë„í‹°ë¥¼ ë¶€ì—¬í•˜ëŠ” L1 ê·œì œë¥¼ ì„ í˜• íšŒê·€ì— ì ìš©í•œ ê²ƒì´ **ë¼ì˜(Lasso)** íšŒê·€ì…ë‹ˆë‹¤.

L1 ê·œì œëŠ” ë¶ˆí•„ìš”í•œ íšŒê·€ ê³„ìˆ˜ë¥¼ ê¸‰ê²©í•˜ê²Œ ê°ì†Œì‹œì¼œ 0ìœ¼ë¡œ ë§Œë“¤ê³  ì œê±°í•©ë‹ˆë‹¤. ì´ëŸ¬í•œ ì ì—ì„œ L1 ê·œì œëŠ” ì ì ˆí•œ í”¼ì²˜ë§Œ íšŒê·€ì— í¬í•¨ì‹œí‚¤ëŠ” í”¼ì²˜ ì„ íƒì˜ íŠ¹ì„±ì„ ê°€ì§€ê³  ìˆìŠµë‹ˆë‹¤. 

ì‚¬ì´í‚·ëŸ°ì€ **Lasso** í´ë˜ìŠ¤ë¥¼ í†µí•´ ë¼ì˜ íšŒê·€ë¥¼ êµ¬í˜„í–ˆìŠµë‹ˆë‹¤. Lasso í´ë˜ìŠ¤ì˜ ì£¼ìš” ìƒì„± íŒŒë¼ë¯¸í„° alphaëŠ” ë¼ì˜ íšŒê·€ì˜ alpha L1 ê·œì œ ê³„ìˆ˜ì— í•´ë‹¹í•©ë‹ˆë‹¤.

í­ê·„ ëª¸ë¬´ê²Œ ì˜ˆì¸¡ ì˜ˆì œë¥¼ Lasso í´ë˜ìŠ¤ë¥¼ ì´ìš©í•´ ë‹¤ì‹œ ì˜ˆì¸¡í•´ë³´ê³ ,

```python
# alpha ê°’ì˜ ë³€í™”ì— ë”°ë¥¸ í‰ê·  RMSEì˜ ë³€í™”

from sklearn.linear_model import Lasso
from sklearn.model_selection import cross_val_score

print('####### Lasso #######\n')

lasso_alphas = [0.07, 0.1, 0.5, 1, 3]

for alpha in lasso_alphas :
    lasso = Lasso(alpha = alpha)

    neg_mse_scores = cross_val_score(lasso, feature, target, scoring="neg_mean_squared_error", cv = 5)
    avg_rmse = np.mean(np.sqrt(-1 * neg_mse_scores))
    print('alpha {0} ì¼ ë•Œ 5 folds ì˜ í‰ê·  RMSE : {1:.3f} '.format(alpha, avg_rmse))
```

### **3) ì—˜ë¼ìŠ¤í‹±ë„· íšŒê·€ (Elastic Net)**


**ì—˜ë¼ìŠ¤í‹±ë„·(Elastic Net)** íšŒê·€ëŠ” L2 ê·œì œì™€ L1 ê·œì œë¥¼ ê²°í•©í•œ íšŒê·€ì…ë‹ˆë‹¤.

ë¼ì˜ íšŒê·€ëŠ” ì„œë¡œ ìƒê´€ê´€ê³„ê°€ ë†’ì€ í”¼ì²˜ë“¤ì˜ ê²½ìš° ì¤‘ìš” í”¼ì²˜ë§Œì„ ì„ íƒí•˜ê³  ë‹¤ë¥¸ í”¼ì²˜ë“¤ì˜ íšŒê·€ ê³„ìˆ˜ëŠ” ëª¨ë‘ 0ìœ¼ë¡œ ë§Œë“œëŠ” ì„±í–¥ì´ ê°•í•©ë‹ˆë‹¤. í•˜ì§€ë§Œ ì´ëŸ¬í•œ ì„±í–¥ìœ¼ë¡œ alpha ê°’ì— ë”°ë¼ íšŒê·€ ê³„ìˆ˜ì˜ ê°’ì´ ê¸‰ê²©íˆ ë³€ë™í•  ìˆ˜ë„ ìˆëŠ”ë°, ì´ë¥¼ ì™„í™”í•˜ê¸° ìœ„í•´ L2 ê·œì œê°€ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.

í•˜ì§€ë§Œ ì—˜ë¼ìŠ¤í‹±ë„·ì˜ ê·œì œëŠ” L1 ê·œì œì™€ L2 ê·œì œê°€ ê²°í•©ëœ ê·œì œì´ê¸° ë•Œë¬¸ì— ìˆ˜í–‰ì‹œê°„ì´ ìƒëŒ€ì ìœ¼ë¡œ ì˜¤ë˜ ê±¸ë¦°ë‹¤ëŠ” ë‹¨ì ì´ ìˆìŠµë‹ˆë‹¤.

<br>


ì‚¬ì´í‚·ëŸ°ì€ **ElasticNet** í´ë˜ìŠ¤ë¥¼ í†µí•´ì„œ ì—˜ë¼ìŠ¤í‹±ë„· íšŒê·€ë¥¼ êµ¬í˜„í•©ë‹ˆë‹¤. 

ElasticNet í´ë˜ìŠ¤ì˜ ì£¼ìš” ìƒì„± íŒŒë¼ë¯¸í„°ëŠ” *alpha*ì™€ *l1_ratio*ì…ë‹ˆë‹¤. ì´ë•Œ alpha ê°’ì€ Ridgeì™€ Lasso í´ë˜ìŠ¤ì˜ alpha ê°’ê³¼ëŠ” ë‹¤ë¦…ë‹ˆë‹¤. 

<br>

ì—˜ë¼ìŠ¤í‹±ë„·ì˜ ê·œì œëŠ” **a * L1 + b * L2** ë¡œ ì •ì˜ë©ë‹ˆë‹¤. 

ë”°ë¼ì„œ,ã…¤ã…¤**alphaã…¤ã…¤--->ã…¤ã…¤a+b**

ã…¤ã…¤ã…¤ã…¤ã…¤ **l1_ratioã…¤ã…¤--->ã…¤ã…¤a/(a+b)** 

<br>

í­ê·„ ëª¸ë¬´ê²Œ ì˜ˆì¸¡ ì˜ˆì œë¥¼ ElasticNet í´ë˜ìŠ¤ë¥¼ ì´ìš©í•´ ë‹¤ì‹œ ì˜ˆì¸¡í•´ë³´ê³ ,

```python
# alpha ê°’ì˜ ë³€í™”ì— ë”°ë¥¸ í‰ê·  RMSEì˜ ë³€í™”

from sklearn.linear_model import ElasticNet
from sklearn.model_selection import cross_val_score

print('####### ElasticNet #######')

elastic_alphas = [0.07, 0.1, 0.5, 1, 3]

for alpha in elastic_alphas :
    elastic = ElasticNet(alpha = alpha, l1_ratio=0.7)

    neg_mse_scores = cross_val_score(elastic, feature, target, scoring="neg_mean_squared_error", cv = 5)
    avg_rmse = np.mean(np.sqrt(-1 * neg_mse_scores))
    print('alpha {0} ì¼ ë•Œ 5 folds ì˜ í‰ê·  RMSE : {1:.3f} '.format(alpha, avg_rmse))
```
