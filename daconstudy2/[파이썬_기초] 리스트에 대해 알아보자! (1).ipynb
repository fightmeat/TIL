*   **인덱싱 & 리버스 인덱스(reverse index)**


```python
#인덱싱 : 리스트에 있는 값 접근을 위해 상대적인 주소 사용
#리버스 인덱스(reverse index) : -1, -2, ...

cities = ['서울','부산','인천','대구','대전','광주','울산','수원']
print(cities[0])  #인덱싱  => 0번째 원소인 '서울' 출력
print(cities[7])
print(cities[-1]) #젤 끝에 있는 원소 '수원' 출력
print(cities[-2]) #뒤에서 두번째 원소 '울산' 출력
```

*   **슬라이싱**



```python
#슬라이싱 : 리스트의 일부를 잘라 반환
#슬라이싱 방법 => 리스트변수명[시작 인덱스 : 마지막 인덱스]
print(cities[0:3])
print(cities[2:5])
print(cities[:4])          #시작 인덱스 X => 0부터 시작
print(cities[4:])          #끝 인덱스 X => 끝까지 해당
print(cities[:-2])


#전체범위 슬라이싱 : ":" <= [0:], [0:len(리스트변수명)]
#증가값: 리스트변수명[시작 인덱스 : 마지막 인덱스 : 증가값]   (시작, 마지막 인덱스 생략 가능)
print(cities[:])              #인덱스 X => 전체 호출

print(cities[::2])          #2씩 증가하는 해당되는 걸 호출
print(cities[1::2])       #1부터 2씩 증가하는

print(cities[::-1])        #맨 끝에서부터 시작
```
