seaborn은 matplotlib을 기반으로 다양한 색상 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지입니다.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

그래프를 그려보기에 앞서, Seaborn으로 구현할 수 있는 그래프에는 무엇이 있을지 알아볼까요?

Seaborn의 공식 사이트에서는 seaborn으로 그릴 수 있는 그래프를 목적에 따라 분류해놓은 것을 볼 수 있습니다.


*   relation plots : 두 변수의 관계를 나타냄
##### 종류 : relplot, scatterplot, lineplot
*   categorical plots : 범주형 데이터를 대상으로 함
##### 종류 : catplot, stripplot, swarmplot, boxplot, violinplot, boxenplot, pointplot, barplot, countplot
*   distribution plots : 변수들의 분포를 나타냄
##### 종류 : displot, hisplot, kdeplot, ecdfplot, rugplot, distplot	
*   regression plots : 회귀분석 결과를 나타내줌
##### 종류 : Implot, regplot, residplot
*   matrix plots : 변수 간의 관계 정도를 매트릭스로 만들고 색을 입혀 나타냄
##### 종류 : heatmap, clustermap
*   multi-plot grids : 여러 그래프를 한번에 나타냄
##### 종류 : FacetGrid, pairplot, PairGrid, jointplot, JointGrid


모든 그래프를 그릴 때 코드의 기본형은 sns.그래프이름(~) 입니다.

그럼 하나씩 그려볼까요?

# 1. scatter plot

먼저 scatter plot 입니다.

두 변수 간의 관계를 나타내는 그래프이기 때문에 x, y 두 개의 변수값을 넣어주어야 합니다.

tips data set 을 이용하겠습니다.

```python
tips = sns.load_dataset('tips')
tips.head()
```

총 비용 대비 팁 금액의 양을 scatter plot으로 나타내 보겠습니다.

```python
sns.scatterplot(x=tips.total_bill, y=tips.tip)
```

data를 지정해주고, x, y에 column명을 써주셔도 됩니다!

축과 타이틀 같은 레이아웃은 matplotlib의 함수들을 활용하여 설정해줄 수 있습니다.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('total bill and tips', fontsize=15)
plt.show()
```

seaborn 에서는 hue 라는 매개변수를 활용할 수 있는데, hue 에 특정 변수값을 지정해주면, 그 변수값에 따라 데이터들이 다른 색상으로 표현됩니다.

```python
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips)
plt.title('total bill and tips', fontsize=15)
plt.show()
```

style 이라는 매개변수 또한 활용할 수 있는데, style 에 특정 변수값을 지정해주면, 그 변수값에 따라 데이터들이 다른 모양으로 표현됩니다.

```python
sns.scatterplot(x='total_bill', y='tip', style='day', data=tips)
plt.title('total bill and tips', fontsize=15)
plt.show()
```

size 라는 매개변수를 설정해줄 수도 있는데, 변수명을 설정해주면 변수값에 따라 marker의 크기를 나타내줍니다.

sizes 는 구체적인 marker의 크기의 범위를 지정해줄 수 있는데, 튜플을 통해 지정해주었습니다.

```python
sns.scatterplot(x='total_bill', y='tip', size='size', sizes=(10, 200), data=tips)
plt.title('total bill and tips', fontsize=15)
plt.show()
```

그래프의 line이나 marker의 색상 설정은 color, palette 매개변수를 통해 가능합니다.

[seaborn 의 color palette 확인](https://seaborn.pydata.org/tutorial/color_palettes.html)

위의 링크에서 다양한 color palette 코드들을 확인할 수 있습니다.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, color='pink')
plt.title('total bill and tips', fontsize=15)
plt.show()
```

특히 hue를 지정해주었을 때, palette를 함께 지정해주면 color palette 대로 범주별 데이터를 나타낼 수 있습니다.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, palette='PuRd', hue='day')
plt.title('total bill and tips', fontsize=15)
plt.show()
```

# 2. line plot

다음은 line plot 입니다.

두 변수 간의 관계를 나타내는 그래프이기 때문에 x, y 두 개의 변수값을 넣어주어야 합니다.


```python
sns.lineplot(x=[1,2,3,4], y=[5,6,8,12])
plt.show()
```

내장 데이터 flights 를 이용해서도 그려보겠습니다.

```python
flights = sns.load_dataset('flights')
flights.head()
```

년도별로 승객수가 어떻게 변화하는지 봅시다!

```python
sns.lineplot(x='year', y='passengers', data=flights)
plt.show()
```

증가하는 것을 볼 수 있네요..!

그래프상의 선 주위의 음영은 데이터값의 오차(값이 월에 따라 달라지기 때문에)를 나타냅니다.

scatter plot 과 마찬가지로, hue 와 style을 설정해줄 수 있습니다.

```python
sns.lineplot(x='year', y='passengers', data=flights, hue='month')
plt.show()
```

# 3. relplot

relplot 은 scatterplot 이나 lineplot을 출력하는데, 범주에 따라 여러개를 동시에 그려줄 수 있습니다.

kind 매개변수를 통해 scatter, line을 지정해줄 수 있습니다.

tips data set에서 요일별로 총 비용 대비 팁 금액의 양을 나타내보았습니다.

그리고 성별별로 다른 색의 marker로 표현해주었습니다.

```python
sns.relplot(
    x='total_bill', y='tip', data=tips, kind='scatter', col='day', hue='sex'
)
plt.show()
```

요일별로 데이터의 양, 총 금액 대비 팁의 양을 비교해볼 수 있네요!

다음으로는 categorical plots 를 그려보겠습니다.

# 4. count plot

count plot은 범주형 데이터를 대상으로 하고, 범주별로 데이터 수를 나타내주기 때문에 범주형 변수 한 가지만 입력해주면 됩니다.

매개변수 x에 변수를 지정하면 수직의 그래프가, y에 변수를 지정하면 수평의 그래프가 출력됩니다!

이번에도 tips data를 이용해보겠습니다.

time 별로 데이터 개수를 나타낸 그래프입니다!

```python
sns.countplot(x='time', data=tips)
plt.show()
```

```python
sns.countplot(y='time', data=tips)
plt.show()
```

마찬가지로, hue 를 지정해줄 수도 있습니다!

```python
sns.countplot(x='time', data=tips, hue='sex')
plt.show()
```

# 5. bar plot

bar plot 은 우리가 익숙한 막대그래프를 생각하면 됩니다.

x와 y, 두 개의 변수를 넣어주어야 하는데, x축에 범주형 데이터 변수를 넣어줄 경우 수직의 막대그래프가, y에 범주형 데이터 변수를 넣어줄 경우 수평의 막대그래프가 형성됩니다.

앞의 countplot과 같이 tips 데이터를 이용해주겠습니다.

```python
sns.barplot(x='time', y='tip', data=tips)
plt.show()
```

```python
sns.barplot(y='time', x='tip', data=tips)
plt.show()
```

막대 그래프 값은 가로축의 범주별로 세로축의 수치값의 평균값을 나타내고 가운데의 검은 선은 오차 범위를 나타냅니다.


# 6. box plot

box plot은 범주별로 데이터 값을 주요 요약 통계량과 함께 시각화해줍니다.

변수 값을 하나만 입력할 경우 수치형 데이터를 입력해야 하고, 수치형 데이터에 대한 box plot을 그려줍니다. 위의 countplot과 마찬가지로, x에 설정하면 수평의, y에 설정하면 수직의 boxplot을 그려줍니다.

변수 값을 x, y 두 개 입력할 경우 barplot과 같이 입력한 두 변수에 대해 각각 가로, 세로축의 데이터로 그래프를 그려줍니다. 그리고 범주형 데이터가 x에 오면 수직의, y에 오면 수평의 boxplot이 출력됩니다!

```python
sns.boxplot(x='tip', data=tips)
plt.show()
```

tip의 범위, 4분위 수, outlier 등을 알 수 있네요.

```python
sns.boxplot(y='tip', data=tips)
plt.show()
```

y에 지정해주었기 때문에 수직의 그래프가 출력되었습니다.

```python
sns.boxplot(x='day', y='tip', data=tips, palette='pastel')
plt.show()
```

범주형 데이터인 day가 x에 설정되었기 때문에 수직의 boxplot이 출력되었습니다.

```python
sns.boxplot(y='day', x='tip', data=tips,  palette='Blues')
plt.show()
```

이번엔 범주형 데이터인 day가 y에 설정되었기 때문에 수평의 boxplot이 출력되었습니다.

# 7. boxen plot

boxen plot은 box plot과 거의 유사하나, 사분위 값마다 데이터의 개수를 박스 크기로 표현해준다는 점이 다르다고 할 수 있습니다.

이 그래프를 처음보시는 분들도 많으실텐데.. 그만큼 잘 쓰이지는 않는 것 같아요.

위의 boxplot과 비교가 가능하도록 같은 데이터로 그려보았습니다.

```python
sns.boxenplot(x='day', y='tip', data=tips, palette='pastel')
plt.show()
```

# 8. violinplot

violin plot 은 box plot 과 비슷하게 전체 데이터 값을 나타냄과 동시에 카테고리별로 각 데이터값의 전체적인 분포를 알 수 있습니다.

boxplot과 마찬가지로 변수 값을 x, y 두 개 입력하고, 입력한 두 변수에 대해 각각 가로, 세로축의 데이터로 그래프를 그려줍니다. 그리고 범주형 데이터가 x에 오면 수직의, y에 오면 수평의 violinplot이 출력됩니다!

이번엔 iris data set을 이용해보겠습니다!

```python
iris = sns.load_dataset('iris')
iris.head()
```

종에 따라 꽃받침의 길이(sepal length)의 값를 비교해볼까요?

```python
sns.violinplot(x='species', y='sepal_length', data=iris, palette='Purples')
plt.show()
```

```python
sns.violinplot(y='species', x='sepal_length', data=iris, palette='Purples')
plt.show()
```

# 9. strip plot

strip plot은 범주별로 데이터 값이 어느 범위에 분포하는지 점으로 나타낸 그래프입니다.

한 점 당 한개의 데이터 값이 대응한다고 생각하면 됩니다.

데이터 값들이 어느 수치에 몰려있는지 확인할 수 있습니다.

위의 violin plot을 그린 데이터를 그대로 이용하겠습니다.

```python
sns.stripplot(x='species', y='sepal_length', data=iris, palette='Purples')
plt.show()
```

마찬가지로, 축변수 설정을 바꿔주면 수평방향의 그래프를 도출할 수 있습니다.

```python
sns.stripplot(y='species', x='sepal_length', data=iris, palette='Purples')
plt.show()
```

# 10. swarm plot

swarm plot은 strip plot와 표현 및 기능이 거의 똑같지만, 데이터의 전체적인 분포를 나타내준다는 점을 더한 그래프입니다.

마찬가지로, 축변수 설정을 바꿔주면 수평방향의 그래프를 도출할 수 있습니다.


```python
sns.swarmplot(x='species', y='sepal_length', data=iris, palette='Purples')
plt.show()
```

scatter로 이루어진 것을 빼면, violin plot과 매우 비슷한 모양이죠?

따라서, violin plot과 겹쳐서 점의 색을 흰색으로 바꾸어 다음과 같이 표현할 수도 있습니다.

```python
sns.violinplot(x='species', y='sepal_length', data=iris, palette='Purples')
sns.swarmplot(x='species', y='sepal_length', data=iris, color='white')
plt.show()
```

# 11. point plot

point plot은 line plot의 오차범위를 막대로 표현해준 그래프라고 생각하면 됩니다.

중앙값을 점 찍고 오차범위를 막대로 표현하여 연결한 그래프인데요.

역시 두개의 변수를 받고, 가로축 변수 x에는 범주형 변수를, 세로축 변수 y에는 수치형 데이터값을 넣어주어야 합니다!

```python
sns.pointplot(x='year', y='passengers', data=flights)
plt.show()
```

```python
sns.lineplot(x='year', y='passengers', data=flights)
plt.show()
```

비교해보니 그래프 표현 방식만 다를 뿐, 정말 동일하죠?

# 12. cat plot

마지막으로 cat plot 입니다.

cat plot은 앞선 rel plot과 카테고리형 그래프 코드로서 동일한 기능을 합니다.

마찬가지로, 여러 그래프를 동시에 나타낼 수 있고, kind를 통해 그래프 형태를 지정해줄 수 있습니다! 

tip data set으로는 요일별 식사 시간대에 따른 남녀의 팁의 양을 다음과 같이 나타내볼 수 있겠네요!

```python
sns.catplot(
    x='time', y='tip', data=tips, kind='bar', col='day', hue='sex'
)
plt.show()
```
