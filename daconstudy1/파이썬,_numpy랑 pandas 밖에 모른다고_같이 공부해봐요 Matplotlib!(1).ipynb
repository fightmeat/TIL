## Matplotlib 란?
데이터 분석을 할 때 데이터 분석 결과를 쉽게 이해할 수 있도록 제시하는 것이 정말 중요한데요. 이러한 데이터 시각화를 하기 위해 가장 많이 이용되는 패키지가 바로 Matplotlib라고 합니다. 
주로 plot을 그리는데 활용되고, 주로 2차원 그림을 위한 패키지를 제공합니다.

## 사용해보자!

plot을 그릴 때 사용하는 함수가 matplotlib의 pyplot 인데요, import 부터 시작합시다!

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

먼저, plt.figure() 는 빈 도화지의 크기를 지정해주는 함수입니다. figsize 인자를 통해, 구체적인 크기를 지정해줄 수 있습니다.

plt.show() 는 그림을 프린트해주는 기능을 합니다.



*   plt.figure(figsize=(m,n)) : 도화지 크기 설정
*   plt.show() : 그림 프린트



먼저 선그래프부터 그려볼까요? 선그래프는 plot() 함수에 X좌표와 Y좌표 값이 담긴 array 혹은 list 를 적용하면 됩니다. 만약 좌표 값을 하나만 제시한다면, Y값으로 인식하고 X좌표 값은 [0,1,2,3 ...]으로 인식한다는 점!도 알고 계시면 편리할거에요!

```python
plt.figure(figsize=(5,3))
plt.plot([1,2,3,4],[2,4,6,8])
plt.show()
```

간단하죠?
이제 제목, 축을 변경해봅시다!
 다음은 제목이나 축을 변경하는 함수들입니다.


*   title() : 그래프 제목 추가 함수
*   xlabel() : x축 레이블 추가 함수
*   ylabel() : y축 레이블 추가 함수
*   xticks() : x축 눈금 표시 함수(labels로 지정 가능)
*   yticks() : y축 눈금 표시 함수(labels로 지정 가능)
*   axis() : x, y축 범위 지정 


```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.figure(figsize=(5,3))
plt.plot(x, y)
plt.title('line plot example')  #제목 추가
plt.xlabel('x label')  #x축 레이블 추가
plt.ylabel('y label')  #y축 레이블 추가
plt.xticks(x, labels=['a','b','c','d'])  #x축 눈금 표시
plt.yticks(y, labels=[2,4,6,8])  #y축 눈금 표시
plt.axis([0,5,0,10])  #x, y축 범위 지정([xmin, xmax, ymin, ymax]의 형태로 넣어주면 됨!)
plt.show()
```

plot() 내의 다양한 인자를 통해 선스타일, 선 색상, 마커(marker)를 설정할 수 있습니다.



선 종류 

|code|type|
|:---:|:---:|
|'-'|solid line|
|'--'|dashed line|
|'-.'|dash-dot line|
|':'|dotted line|


색상

|code|type|
|---|:---:|
|'b' or 'blue'|blue|
|'g' or 'green'|green|
|'y' or 'yellow'|yellow|
|'k' or 'black'|black|
|'w' or 'white'|white|


마커

|code|type|
|:---:|:---:|
|'.'|점 모양|
|'o'|원 모양|
|'v'|삼각형 모양|
|'s'|정사각형 모양|
|'*'|별 모양|
|'+'|+ 모양|

```python
plt.plot(figsize=(5,3))
plt.plot([1,5,8,15], linestyle=':', color='red', marker='o')
plt.title('line plot example')
plt.show()
```

함수 내에 인자를 쓰지 않고, '색상-마커-선종류' 순으로 묶어서 나타낼 수도 있습니다.

```python
plt.plot(figsize=(5,3))
plt.plot([1,5,8,15], 'ro:') #색상 red, 원 모양 마커, 점선 적용
plt.title('line plot example')
plt.show()
```

이번엔 범례를 추가해보겠습니다.

범례는 legend() 함수를 이용하면 되는데요,  plot 함수 내에 label 인자를 지정해주면  범례로 인식합니다.

```python
x1=np.random.randint(1,10,10)
x2=np.random.randint(1,10,10)
plt.plot(x1, label='x1')
plt.plot(x2, label='x2')
plt.title('random numbers(from 1 to 10)')
plt.legend()
plt.show()
```

loc 인자를 통해, 범례의 위치를 설정해 줄 수도 있습니다.

또한 ncol 인자를 통해, 범례의 열의 개수를 설정해 줄 수도 있습니다.

```python
x1=np.random.randint(1,10,10)
x2=np.random.randint(1,10,10)
plt.plot(x1, label='x1')
plt.plot(x2, label='x2')
plt.title('random numbers(from 1 to 10)')
plt.legend(loc=(0.5, 0.7), ncol=2)  #loc : x축, y축을 1이라고 할 때, 좌표로 위치를 표현한다고 생각!, ncol : 디폴트는 1이다!
plt.show()
```
